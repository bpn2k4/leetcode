```python
from typing import List


class Solution:
  def uniqueXorTriplets(self, nums: List[int]) -> int:
    n = len(nums)
    if n <= 2:
      return n
    ans = 1
    while ans <= n:
      ans <<= 1
    return ans
```

```python
nums = [1, 2]
Solution().uniqueXorTriplets(nums=nums)
```

```python
nums = [3, 1, 2]
Solution().uniqueXorTriplets(nums=nums)
```
