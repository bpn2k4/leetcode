```python
from typing import List
from collections import Counter


class Solution:
  def maximumLength(self, nums: List[int]) -> int:
    num2count = Counter(nums)
    max_length = max(1, (num2count.get(1, 0) - 1) // 2 * 2 + 1)
    if 1 in num2count: del num2count[1]
    for num in num2count.keys():
      if num2count.get(num) < 2:
        continue
      length = 2
      num = num * num
      while num in num2count and num2count.get(num) >= 2:
        length += 2
        num = num * num
      if num2count.get(num, 0) == 1:
        length += 1
        max_length = max(max_length, length)
      else:
        length -= 1
        max_length = max(max_length, length)
    return max_length
```
