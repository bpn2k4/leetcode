```python
from typing import List


class Solution:
  def prefixesDivBy5(self, nums: List[int]) -> List[bool]:
    
    result = []
    prefix = 0
    for num in nums:
      prefix = (prefix * 2 + num) % 5
      result.append(prefix == 0)
    
    return result
```

```python
nums = [0, 1, 1]
Solution().prefixesDivBy5(nums=nums)
```

```python
nums = [1, 1, 1]
Solution().prefixesDivBy5(nums=nums)
```
