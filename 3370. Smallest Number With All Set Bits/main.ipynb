```python
class Solution:
  def smallestNumber(self, n: int) -> int:
    
    binary = bin(n)[2:].replace('0', '1')
    
    return int(binary, 2)
```

```python
n = 5
Solution().smallestNumber(n=n)
```

```python
n = 10
Solution().smallestNumber(n=n)
```
