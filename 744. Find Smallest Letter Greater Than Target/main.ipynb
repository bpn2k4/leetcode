```python
from typing import List


class Solution:
  def nextGreatestLetter(self, letters: List[str], target: str) -> str:
    for char in letters:
      if ord(char) > ord(target):
        return char
    return letters[0]
```

```python
letters = ["c", "f", "j"]
target = "a"
Solution().nextGreatestLetter(letters=letters, target=target)
```

```python
letters = ["c", "f", "j"]
target = "c"
Solution().nextGreatestLetter(letters=letters, target=target)
```

```python
letters = ["x", "x", "y", "y"]
target = "z"
Solution().nextGreatestLetter(letters=letters, target=target)
```
