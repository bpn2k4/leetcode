```python
from typing import List


class Solution:
  def maximumElementAfterDecrementingAndRearranging(self, arr: List[int]) -> int:
    arr.sort()
    if arr[0] != 1:
      arr[0] = 1
    for i in range(1, len(arr)):
      if arr[i] >= arr[i - 1] + 1:
        arr[i] = arr[i - 1] + 1
    return arr[-1]
```

```python
arr = [2, 2, 1, 2, 1]
Solution().maximumElementAfterDecrementingAndRearranging(arr=arr)
```

```python
arr = [100, 1, 1000]
Solution().maximumElementAfterDecrementingAndRearranging(arr=arr)
```
