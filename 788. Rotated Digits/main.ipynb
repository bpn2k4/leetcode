```python
class Solution:
  def is_good_number(self, num: int) -> bool:
    s = str(num)
    invalid_digits = {'3', '4', '7'}
    changed_digits = {'2', '5', '6', '9'}
    if any(digit in invalid_digits for digit in s):
      return False
    return any(digit in changed_digits for digit in s)

  def rotatedDigits(self, n: int) -> int:
    count = 0
    for number in range(1, n + 1):
      if self.is_good_number(number):
        count += 1
    return count
```

```python
n = 10
Solution().rotatedDigits(n=n)
```

```python
n = 1
Solution().rotatedDigits(n=n)
```
