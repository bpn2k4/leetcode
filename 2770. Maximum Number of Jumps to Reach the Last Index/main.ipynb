```python
from typing import List


class Solution:
  def maximumJumps(self, nums: List[int], target: int) -> int:
    n = len(nums)
    dp = [float("-inf")] * n
    dp[0] = 0
    for i in range(1, n):
      for j in range(i):
        if abs(nums[j] - nums[i]) <= target:
          dp[i] = max(dp[i], dp[j] + 1)
    return -1 if dp[n - 1] < 0 else dp[n - 1]
```

```python
nums = [1, 3, 6, 4, 1, 2]
target = 2
Solution().maximumJumps(nums=nums, target=target)
```
