```python
from typing import List


class Solution:
  def maxDistance(self, nums1: List[int], nums2: List[int]) -> int:
    i = j = 0
    max_distance = 0
    while i < len(nums1) and j < len(nums2):
      if nums1[i] <= nums2[j]:
        max_distance = max(max_distance, j - i)
        j += 1
      else:
        i += 1
    return max_distance
```

```python
nums1 = [55, 30, 5, 4, 2]
nums2 = [100, 20, 10, 10, 5]
Solution().maxDistance(nums1=nums1, nums2=nums2)
```

```python
nums1 = [2, 2, 2]
nums2 = [10, 10, 1]
Solution().maxDistance(nums1=nums1, nums2=nums2)
```
