```python
class Solution:
  def addBinary(self, a: str, b: str) -> str:
    if len(a) > len(b):
      a, b = b, a
    a = [int(char) for char in a]
    b = [int(char) for char in b]
    a = [0] * (len(b) - len(a)) + a
    r = 0
    for i in range(len(b) - 1, -1, -1):
      c = a[i] + b[i] + r
      if c == 3:
        a[i] = 1
        r = 1
      elif c == 2:
        a[i] = 0
        r = 1
      else:
        a[i] = c
        r = 0
    if r == 1:
      a = [1] + a
    return ''.join([str(char) for char in a])
```

```python
a = "11"
b = "1"
Solution().addBinary(a=a, b=b)
```

```python
a = "1010"
b = "1011"
Solution().addBinary(a=a, b=b)
```
