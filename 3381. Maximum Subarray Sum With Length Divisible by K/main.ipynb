```python
from typing import List


class Solution:
  def maxSubarraySum(self, nums: List[int], k: int) -> int:

    prefix = 0
    best = [float('inf')] * k
    best[0] = 0
    answer = -float('inf')
    for i, num in enumerate(nums, start=1):
      prefix += num
      m = i % k

      if best[m] != float('inf'):
        answer = max(answer, prefix - best[m])

      best[m] = min(best[m], prefix)

    return answer
```

```python
nums = [1, 2]
k = 1
Solution().maxSubarraySum(nums=nums, k=k)
```

```python
nums = [-1, -2, -3, -4, -5]
k = 4
Solution().maxSubarraySum(nums=nums, k=k)
```

```python
nums = [-5, 1, 2, -3, 4]
k = 2
Solution().maxSubarraySum(nums=nums, k=k)
```
