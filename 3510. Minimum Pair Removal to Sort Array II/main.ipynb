```python
from typing import List, Optional
import heapq


class Node:
  __slots__ = ("val", "prev", "next", "alive", "order")

  def __init__(self, val: int, order: int):
    self.val: int = val
    self.prev: Optional[Node] = None
    self.next: Optional[Node] = None
    self.alive: bool = True
    self.order: int = order


class Solution:
  def minimumPairRemoval(self, nums: List[int]) -> int:
    n = len(nums)
    if n <= 1:
      return 0

    nodes = [Node(nums[i], i) for i in range(n)]
    for i in range(n - 1):
      nodes[i].next = nodes[i + 1]
      nodes[i + 1].prev = nodes[i]

    bad = 0
    for i in range(n - 1):
      if nums[i] > nums[i + 1]:
        bad += 1
    if bad == 0:
      return 0

    heap: List[tuple[int, int, Node]] = []
    for i in range(n - 1):
      heapq.heappush(heap, (nodes[i].val + nodes[i + 1].val, nodes[i].order, nodes[i]))

    ops = 0

    while bad > 0:
      s, _, left = heapq.heappop(heap)

      if not left.alive or not left.next or not left.next.alive:
        continue
      if left.val + left.next.val != s:
        continue

      right = left.next

      if left.prev and left.prev.val > left.val:
        bad -= 1
      if left.val > right.val:
        bad -= 1
      if right.next and right.val > right.next.val:
        bad -= 1

      left.val += right.val
      right.alive = False
      left.next = right.next
      if right.next:
        right.next.prev = left

      if left.prev and left.prev.val > left.val:
        bad += 1
      if left.next and left.val > left.next.val:
        bad += 1

      if left.prev:
        heapq.heappush(
            heap,
            (left.prev.val + left.val, left.prev.order, left.prev)
        )
      if left.next:
        heapq.heappush(
            heap,
            (left.val + left.next.val, left.order, left)
        )

      ops += 1

    return ops
```

```python
nums = [5, 2, 3, 1]
Solution().minimumPairRemoval(nums=nums)
```
