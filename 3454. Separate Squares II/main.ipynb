```python
from typing import List


class Solution:
  def separateSquares(self, squares: List[List[int]]) -> float:
    events = []
    for x, y, l in squares:
      events.append((y, 1, x, x + l))
      events.append((y + l, -1, x, x + l))

    events.sort()

    def union_length(intervals):
      if not intervals:
        return 0.0
      intervals.sort()
      total = 0.0
      cur_start, cur_end = intervals[0]
      for s, e in intervals[1:]:
        if s > cur_end:
          total += cur_end - cur_start
          cur_start, cur_end = s, e
        else:
          cur_end = max(cur_end, e)
      total += cur_end - cur_start
      return total

    active = []
    prev_y = events[0][0]
    area_segments = []
    total_area = 0.0

    for y, typ, x1, x2 in events:
      if y > prev_y:
        width = union_length(active)
        height = y - prev_y
        if width > 0:
          area = width * height
          area_segments.append((prev_y, y, width))
          total_area += area
        prev_y = y

      if typ == 1:
        active.append((x1, x2))
      else:
        active.remove((x1, x2))

    half = total_area / 2.0
    acc = 0.0

    for y1, y2, width in area_segments:
      slab_area = width * (y2 - y1)
      if acc + slab_area >= half:
        return y1 + (half - acc) / width
      acc += slab_area

    return prev_y
```

```python
squares = [[0, 0, 1], [2, 2, 1]]
Solution().separateSquares(squares=squares)
```

```python
squares = [[0, 0, 2], [1, 1, 1]]
Solution().separateSquares(squares=squares)
```
