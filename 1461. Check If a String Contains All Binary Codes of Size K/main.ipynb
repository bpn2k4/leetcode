```python
class Solution:
  def hasAllCodes(self, s: str, k: int) -> bool:
    n = len(s)
    if n < k:
      return False

    total_needed = 1 << k
    seen = set()
    current = 0
    mask = total_needed - 1
    for i in range(n):
      current = ((current << 1) & mask) | int(s[i])
      if i >= k - 1:
        seen.add(current)
        if len(seen) == total_needed:
          return True
    return False
```

```python
s = "00110110"
k = 2
Solution().hasAllCodes(s=s, k=k)
```
