```python
from typing import List


class Solution:
  def constructTransformedArray(self, nums: List[int]) -> List[int]:
    n = len(nums)
    result = [0] * n
    for i in range(n):
      result[i] = nums[(i + nums[i]) % n]
    return result
```

```python
nums = [3, -2, 1, 1]
Solution().constructTransformedArray(nums=nums)
```

```python
nums = [-1, 4, -1]
Solution().constructTransformedArray(nums=nums)
```
