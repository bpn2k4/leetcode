```python
from typing import List


class Solution:
  def minOperations(self, nums: List[int]) -> int:

    s = []
    res = 0
    for a in nums:
      while s and s[-1] > a:
        s.pop()
      if a == 0:
        continue
      if not s or s[-1] < a:
        res += 1
        s.append(a)

    return res
```

```python
nums = [0, 2]
Solution().minOperations(nums=nums)
```

```python
nums = [3, 1, 2, 1]
Solution().minOperations(nums=nums)
```

```python
nums = [1, 2, 1, 2, 1, 2]
Solution().minOperations(nums=nums)
```
