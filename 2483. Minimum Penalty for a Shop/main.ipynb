```python
class Solution:
  def bestClosingTime(self, customers: str) -> int:
    
    n = len(customers)
    penalty = customers.count('Y')
    hour2penalty = [penalty] * (n + 1)
    for i in range(1, n + 1):
      if customers[i - 1] == 'Y':
        penalty -= 1
      if customers[i - 1] == 'N':
        penalty += 1
      hour2penalty[i] = penalty
    
    return hour2penalty.index(min(hour2penalty))
```

```python
customers = "YYNY"
Solution().bestClosingTime(customers=customers)
```

```python
customers = "NNNNN"
Solution().bestClosingTime(customers=customers)
```
