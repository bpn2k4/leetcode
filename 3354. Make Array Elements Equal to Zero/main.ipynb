```python
from typing import List


class Solution:
  def countValidSelections(self, nums: List[int]) -> int:

    n = len(nums)
    ans = 0
    s = sum(nums)
    left, right = 0, s
    for i in range(n):
      if nums[i] == 0:
        if 0 <= left - right <= 1:
          ans += 1
        if 0 <= right - left <= 1:
          ans += 1
      else:
        left += nums[i]
        right -= nums[i]

    return ans
```

```python
nums = [1, 0, 2, 0, 3]
Solution().countValidSelections(nums=nums)
```

```python
nums = [2, 3, 4, 0, 4, 1, 0]
Solution().countValidSelections(nums=nums)
```
