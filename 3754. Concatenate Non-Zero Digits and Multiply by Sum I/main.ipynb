```python
class Solution:
  def sumAndMultiply(self, n: int) -> int:
    s = 0
    num = '0'
    for char in str(n):
      if char != '0':
        s += int(char)
        num += char
    return s * int(num)
```

```python
n = 10203004
Solution().sumAndMultiply(n=n)
```

```python
n = 1000
Solution().sumAndMultiply(n=n)
```
