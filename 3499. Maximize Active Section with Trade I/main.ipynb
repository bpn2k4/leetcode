```python
class Solution:
  def maxActiveSectionsAfterTrade(self, s: str) -> int:
    count_one = s.count('1')
    n = len(s)
    i = 0
    best_gain = 0
    prev = -1000000000
    while i < n:
      start = i
      while i < n and s[i] == s[start]:
        i += 1
      if s[start] == '0':
        cur = i - start
        best_gain = max(best_gain, prev + cur)
        prev = cur
    return count_one + best_gain
```

```python
s = "01"
Solution().maxActiveSectionsAfterTrade(s=s)
```
