```python
from typing import List


class Solution:
  def maxDistance(self, colors: List[int]) -> int:
    n = len(colors)
    max_distance = 0
    for i in range(n):
      for j in range(i + 1, n):
        if colors[i] != colors[j]:
          max_distance = max(max_distance, j - i)
    return max_distance
```

```python
colors = [1, 1, 1, 6, 1, 1, 1]
Solution().maxDistance(colors=colors)
```

```python
colors = [1, 8, 3, 8, 3]
Solution().maxDistance(colors=colors)
```

```python
colors = [0, 1]
Solution().maxDistance(colors=colors)
```
