```python
from typing import List


class Solution:
  def pyramidTransition(self, bottom: str, allowed: List[str]) -> bool:
    
    bottom2candidates = {}
    for allow in allowed:
      if allow[:2] not in bottom2candidates:
        bottom2candidates[allow[:2]] = []
      bottom2candidates[allow[:2]].append(allow[2])
    
    n = len(bottom)
    arr = [list(bottom)]
    for i in range(1, n):
      arr.append(['' for _ in range(n - i)])

    find = [False]

    def TRY(i: int, j: int):
      if find[0]:
        return
      b = arr[i - 1][j] + arr[i - 1][j + 1]
      for candidate in bottom2candidates.get(b, []):
        arr[i][j] = candidate
        if i == n - 1 and j == n - i - 1:
          find[0] = True
        else:
          if j == n - i - 1:
            TRY(i + 1, 0)
          else:
            TRY(i, j + 1)
        arr[i][j] = ''
    
    TRY(1, 0)
    return find[0]
```

```python
bottom = "BCD"
allowed = ["BCC", "CDE", "CEA", "FFF"]
Solution().pyramidTransition(bottom=bottom, allowed=allowed)
```
