```python
from typing import List


class Solution:
  def maxProductPath(self, grid: List[List[int]]) -> int:
    MOD = 10**9 + 7
    m, n = len(grid), len(grid[0])
    pos_dp = [[0 for _ in range(n)] for _ in range(m)]
    neg_dp = [[0 for _ in range(n)] for _ in range(m)]
    pos_dp[0][0] = neg_dp[0][0] = grid[0][0]
    for i in range(1, m):
      pos_dp[i][0] = neg_dp[i][0] = pos_dp[i - 1][0] * grid[i][0]
    for j in range(1, n):
      pos_dp[0][j] = neg_dp[0][j] = pos_dp[0][j - 1] * grid[0][j]
    for i in range(1, m):
      for j in range(1, n):
        if grid[i][j] >= 0:
          pos_dp[i][j] = max(pos_dp[i][j - 1], pos_dp[i - 1][j]) * grid[i][j]
          neg_dp[i][j] = min(neg_dp[i][j - 1], neg_dp[i - 1][j]) * grid[i][j]
        else:
          pos_dp[i][j] = min(neg_dp[i][j - 1], neg_dp[i - 1][j]) * grid[i][j]
          neg_dp[i][j] = max(pos_dp[i][j - 1], pos_dp[i - 1][j]) * grid[i][j]
    if pos_dp[m - 1][n - 1] < 0:
      return -1
    return pos_dp[m - 1][n - 1] % MOD
```

```python
grid = [[-1, -2, -3], [-2, -3, -3], [-3, -3, -2]]
Solution().maxProductPath(grid=grid)
```

```python
grid = [[1, -2, 1], [1, -2, 1], [3, -4, 1]]
Solution().maxProductPath(grid=grid)
```
