```python
from collections import Counter


class Solution:
  def numberOfSpecialChars(self, word: str) -> int:
    char2count = Counter(word)
    count = 0
    for char in char2count:
      if char.islower() and char.upper() in char2count:
        count += 1
    return count
```

```python
word = "aaAbcBC"
Solution().numberOfSpecialChars(word=word)
```

```python
word = "abc"
Solution().numberOfSpecialChars(word=word)
```

```python
word = "abBCab"
Solution().numberOfSpecialChars(word=word)
```
