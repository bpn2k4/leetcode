```python
from typing import List


class Solution:
  def minOperations(self, nums: List[int], k: int) -> int:
    
    return sum(nums) % k
```

```python
nums = [3, 9, 7]
k = 5
Solution().minOperations(nums=nums, k=k)
```

```python
nums = [4, 1, 3]
k = 4
Solution().minOperations(nums=nums, k=k)
```

```python
nums = [3, 2]
k = 6
Solution().minOperations(nums=nums, k=k)
```
