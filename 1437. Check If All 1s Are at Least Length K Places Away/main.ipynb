```python
from typing import List


class Solution:
  def kLengthApart(self, nums: List[int], k: int) -> bool:
    
    distance = -1
    for num in nums:
      if num == 1:
        if distance == -1:
          distance = 0
        else:
          if distance < k:
            return False
          distance = 0
      else:
        if distance != -1:
          distance += 1
    
    return True
```

```python
nums = [1, 0, 0, 0, 1, 0, 0, 1]
k = 2
Solution().kLengthApart(nums=nums, k=k)
```
