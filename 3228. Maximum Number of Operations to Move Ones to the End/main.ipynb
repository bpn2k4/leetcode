```python
class Solution:
  def maxOperations(self, s: str) -> int:
    
    n = len(s)
    operations = 0
    count_one = 0
    for i in range(n):
      if s[i] == '1':
        count_one += 1
      else:
        if i + 1 < n and s[i + 1] == '1':
          operations += count_one
    if s[-1] == '0':
      operations += count_one

    return operations
```

```python
s = "1001101"
Solution().maxOperations(s=s)
```

```python
s = "00111"
Solution().maxOperations(s=s)
```
