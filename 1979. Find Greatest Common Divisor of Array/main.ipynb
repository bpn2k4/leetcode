```python
from typing import List
import math


class Solution:
  def findGCD(self, nums: List[int]) -> int:
    return math.gcd(min(nums), max(nums))
```

```python
nums = [2, 5, 6, 9, 10]
Solution().findGCD(nums=nums)
```

```python
nums = [7, 5, 6, 8, 3]
Solution().findGCD(nums=nums)
```

```python
nums = [3, 3]
Solution().findGCD(nums=nums)
```
