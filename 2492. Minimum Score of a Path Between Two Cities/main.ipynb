```python
from typing import List


class Solution:
  def minScore(self, n: int, roads: List[List[int]]) -> int:
    root = list(range(n + 1))

    def find(i: int) -> int:
      root[i] = find(root[i]) if root[i] != i else i
      return root[i]

    for x, y, _ in roads:
      root[find(x)] = find(y)
    res, g1 = 10001, find(1)
    for x, _, d in roads:
      if find(x) == g1:
        res = min(res, d)
    return res
```

```python
n = 4
roads = [[1, 2, 9], [2, 3, 6], [2, 4, 5], [1, 4, 7]]
Solution().minScore(n=n, roads=roads)
```
