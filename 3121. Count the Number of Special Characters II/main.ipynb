```python
class Solution:
  def numberOfSpecialChars(self, word: str) -> int:
    char2first_index = {}
    for index, char in enumerate(word):
      if char.isupper() and char not in char2first_index:
        char2first_index[char] = index
    char2last_index = {}
    for index, char in enumerate(word):
      if char.islower():
        char2last_index[char] = index
    count = 0
    for char in char2last_index:
      if char.upper() in char2first_index and char2first_index.get(char.upper(), -1) > char2last_index[char]:
        count += 1
    return count
```

```python
word = "aaAbcBC"
Solution().numberOfSpecialChars(word=word)
```

```python
word = "abc"
Solution().numberOfSpecialChars(word=word)
```
