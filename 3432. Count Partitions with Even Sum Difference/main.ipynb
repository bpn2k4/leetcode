```python
from typing import List
class Solution:
  def countPartitions(self, nums: List[int]) -> int:
    
    sum_all = sum(nums)
    prefix = nums[0]
    count = 0
    for num in nums[1:]:
      prefix += num
      right = sum_all - prefix
      if abs(prefix - right) % 2 == 0:
        count += 1

    return count
```

```python
nums = [10, 10, 3, 7, 6]
Solution().countPartitions(nums=nums)
```

```python
nums = [1, 2, 2]
Solution().countPartitions(nums=nums)
```

```python
nums = [2, 4, 6, 8]
Solution().countPartitions(nums=nums)
```
