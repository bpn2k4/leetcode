```python
from typing import List


class Solution:
  def minNumberOperations(self, target: List[int]) -> int:

    n = len(target)
    ans = target[0]
    for i in range(1, n):
      ans += max(target[i] - target[i - 1], 0)

    return ans
```

```python
target = [1, 2, 3, 2, 1]
Solution().minNumberOperations(target=target)
```

```python
target = [3, 1, 1, 2]
Solution().minNumberOperations(target=target)
```
