```python
from typing import List


class Solution:
  def latestDayToCross(self, row: int, col: int, cells: List[List[int]]) -> int:

    def dfs(i: int, j: int, mark: set) -> bool:
      if i == row and (i, j) not in mark:
        return True
      mark.add((i, j))
      for x, y in [(i + 1, j), (i - 1, j), (i, j + 1), (i, j - 1)]:
        if x < 1 or x > row or y < 1 or y > col or (x, y) in mark:
          continue
        if dfs(x, y, mark):
          return True
      return False
    
    low, high = 0, len(cells)
    while low < high:
      mid = (low + high) // 2
      mark = set((i, j) for i, j in cells[:mid + 1])
      if any(dfs(1, j, mark) for j in range(1, col + 1) if (1, j) not in mark):
        low = mid + 1
      else:
        high = mid

    return low
```

```python
row = 2
col = 2
cells = [[1, 1], [2, 1], [1, 2], [2, 2]]
Solution().latestDayToCross(row=row, col=col, cells=cells)
```
