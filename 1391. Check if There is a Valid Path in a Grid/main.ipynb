```python
from typing import List
from collections import deque


class Solution:
  def hasValidPath(self, grid: List[List[int]]) -> bool:
    m, n = len(grid), len(grid[0])
    directions = {
      1: [(0, -1), (0, 1)],
      2: [(-1, 0), (1, 0)],
      3: [(0, -1), (1, 0)],
      4: [(0, 1), (1, 0)],
      5: [(0, -1), (-1, 0)],
      6: [(0, 1), (-1, 0)],
    }
    visited = [[False] * n for _ in range(m)]

    def dfs(i: int, j: int) -> bool:
      if visited[i][j]:
        return False
      visited[i][j] = True
      if i == m - 1 and j == n - 1:
        return True
      for di, dj in directions[grid[i][j]]:
        ni, nj = i + di, j + dj
        if 0 <= ni < m and 0 <= nj < n:
          if (-di, -dj) in directions[grid[ni][nj]]:
            if dfs(ni, nj):
              return True
      return False

    return dfs(0, 0)
```

```python
grid = [[2, 4, 3], [6, 5, 2]]
Solution().hasValidPath(grid=grid)
```
