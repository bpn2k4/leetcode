```python
class Solution:
  def canBeEqual(self, s1: str, s2: str) -> bool:
    mark = {}
    for i in range(len(s1)):
      c1, c2 = s1[i], s2[i]
      mark[(c1, i % 2)] = mark.get((c1, i % 2), 0) + 1
      mark[(c2, i % 2)] = mark.get((c2, i % 2), 0) - 1
    return all(v == 0 for v in mark.values())
```

```python
s1 = "abcd"
s2 = "cdab"
Solution().canBeEqual(s1=s1, s2=s2)
```

```python
s1 = "abcd"
s2 = "dacb"
Solution().canBeEqual(s1=s1, s2=s2)
```
