```python
from typing import List


class Solution:
  def maxSideLength(self, mat: List[List[int]], threshold: int) -> int:
    m, n = len(mat), len(mat[0])
    prefix_sum = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
      for j in range(1, n + 1):
        prefix_sum[i][j] = (
            mat[i - 1][j - 1]
            + prefix_sum[i - 1][j]
            + prefix_sum[i][j - 1]
            - prefix_sum[i - 1][j - 1]
        )
    max_side = 0
    for i in range(m):
      for j in range(n):
        max_size = min(m - i, n - j)
        for k in range(1, max_size + 1):
          square_sum = (
              prefix_sum[i + k][j + k]
              - prefix_sum[i][j + k]
              - prefix_sum[i + k][j]
              + prefix_sum[i][j]
          )
          if square_sum <= threshold:
            max_side = max(max_side, k)
    return max_side

  def maxSideLength(self, mat: List[List[int]], threshold: int) -> int:
    m, n = len(mat), len(mat[0])
    prefix_sum = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
      for j in range(1, n + 1):
        prefix_sum[i][j] = (
            mat[i - 1][j - 1]
            + prefix_sum[i - 1][j]
            + prefix_sum[i][j - 1]
            - prefix_sum[i - 1][j - 1]
        )

    def exists_square(k):
      for r in range(m - k + 1):
        for c in range(n - k + 1):
          square_sum = (
              prefix_sum[r + k][c + k]
              - prefix_sum[r][c + k]
              - prefix_sum[r + k][c]
              + prefix_sum[r][c]
          )
          if square_sum <= threshold:
            return True
      return False

    left, right = 1, min(m, n)
    answer = 0
    while left <= right:
      mid = (left + right) // 2
      if exists_square(mid):
        answer = mid
        left = mid + 1
      else:
        right = mid - 1
    return answer
```
