```python
class Solution:
  def angleClock(self, hour: int, minutes: int) -> float:
    if hour == 12:
      hour = 0
    a = hour / 12 * 360 + minutes / 60 * 360 / 12
    b = minutes / 60 * 360
    c = abs(a - b)
    if c > 180:
      c = 360 - c
    return c
```

```python
hour = 12
minutes = 30
Solution().angleClock(hour=hour, minutes=minutes)
```
