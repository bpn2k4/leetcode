```python
from collections import Counter

class Solution:
  def furthestDistanceFromOrigin(self, moves: str) -> int:
    char2count = Counter(moves)
    return char2count.get('_', 0) + abs(char2count.get('L', 0) - char2count.get('R', 0))
```

```python
moves = "L_RL__R"
Solution().furthestDistanceFromOrigin(moves=moves)
```

```python
moves = "_R__LL_"
Solution().furthestDistanceFromOrigin(moves=moves)
```

```python
moves = "_______"
Solution().furthestDistanceFromOrigin(moves=moves)
```
