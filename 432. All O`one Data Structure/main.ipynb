```python
from typing import Optional


class Node:
  def __init__(self):
    self.keys: set[str] = set()
    self.prev: Optional[Node] = None
    self.next: Optional[Node] = None
    self.count: int = 0


class AllOne:
  def __init__(self):
    self.head = Node()
    self.tail = Node()
    self.head.next = self.tail
    self.tail.prev = self.head
    self.key_count: dict[str, int] = {}
    self.count_node: dict[int, Node] = {}

  def _add_node(self, prev_node: Node, new_node: Node):
    new_node.prev = prev_node
    new_node.next = prev_node.next
    prev_node.next.prev = new_node
    prev_node.next = new_node

  def _remove_node(self, node: Node):
    node.prev.next = node.next
    node.next.prev = node.prev

  def _remove_key_from_node(self, node: Node, key: str):
    node.keys.remove(key)
    if not node.keys:
      self._remove_node(node)
      del self.count_node[node.count]

  def inc(self, key: str) -> None:
    if key in self.key_count:
      curr_count = self.key_count[key]
      self.key_count[key] += 1
    else:
      curr_count = 0
      self.key_count[key] = 1

    prev_node = self.count_node.get(curr_count, self.head)
    new_count = curr_count + 1
    if new_count not in self.count_node:
      new_node = Node()
      new_node.count = new_count
      self.count_node[new_count] = new_node
      self._add_node(prev_node, new_node)

    self.count_node[new_count].keys.add(key)

    if curr_count > 0:
      self._remove_key_from_node(prev_node, key)

  def dec(self, key: str) -> None:
    curr_count = self.key_count[key]
    if curr_count == 1:
      del self.key_count[key]
    else:
      self.key_count[key] -= 1

    curr_node = self.count_node[curr_count]
    if curr_count > 1:
      new_count = curr_count - 1
      if new_count not in self.count_node:
        new_node = Node()
        new_node.count = new_count
        self.count_node[new_count] = new_node
        self._add_node(curr_node.prev, new_node)

      self.count_node[new_count].keys.add(key)

    self._remove_key_from_node(curr_node, key)

  def getMaxKey(self) -> str:
    if self.tail.prev == self.head:
      return ""
    return next(iter(self.tail.prev.keys))

  def getMinKey(self) -> str:
    if self.head.next == self.tail:
      return ""
    return next(iter(self.head.next.keys))
```
