```python
class Solution:
  def numberOfWays(self, corridor: str) -> int:
    MOD = 10**9 + 7
    number_of_seat = corridor.count('S')
    if number_of_seat == 0 or number_of_seat % 2 == 1:
      return 0

    count = 1
    seat_index = []
    for i in range(len(corridor)):
      if corridor[i] == 'S':
        seat_index.append(i)
    for i in range(2, len(seat_index), 2):
      p = seat_index[i] - seat_index[i - 1]
      count = count  * p % MOD

    return count
```

```python
corridor = "SSPPSPS"
Solution().numberOfWays(corridor=corridor)
```

```python
corridor = "PPSPSP"
Solution().numberOfWays(corridor=corridor)
```
