```python
from typing import List


class Bank:
  def __init__(self, balance: List[int]):
    self.account2balance = dict(enumerate(balance, start=1))

  def transfer(self, account1: int, account2: int, money: int) -> bool:
    if account1 not in self.account2balance or account2 not in self.account2balance:
      return False

    if account1 == account2:
      balance = self.account2balance[account1]
      if balance < money:
        return False
      return True

    balance1, balance2 = self.account2balance[account1], self.account2balance[account2]
    if balance1 < money:
      return False

    self.account2balance[account1] = balance1 - money
    self.account2balance[account2] = balance2 + money
    return True

  def deposit(self, account: int, money: int) -> bool:
    if account not in self.account2balance:
      return False

    self.account2balance[account] += money
    return True

  def withdraw(self, account: int, money: int) -> bool:
    if account not in self.account2balance:
      return False

    balance = self.account2balance[account]
    if balance < money:
      return False

    self.account2balance[account] = balance - money
    return True
```

```python
bank = Bank([10,100,20,50,30])
print(bank.withdraw(3, 10))
print(bank.transfer(5, 1, 20))
print(bank.deposit(5, 20))
print(bank.transfer(3, 4, 15))
print(bank.withdraw(10, 50))
```

```python
bank = Bank([0])
print(bank.deposit(1, 1000000000000))
print(bank.transfer(1, 1, 1000000000000))
print(bank.withdraw(1, 1000000000000))
print(bank.deposit(1, 0))
print(bank.transfer(1, 1, 0))
print(bank.withdraw(1, 0))
```
