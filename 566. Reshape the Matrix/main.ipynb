```python
from typing import List


class Solution:
  def matrixReshape(self, mat: List[List[int]], r: int, c: int) -> List[List[int]]:
    
    m, n = len(mat), len(mat[0])
    if m * n != r * c:
      return mat

    output = [[0] * c for _ in range(r)]
    for i in range(m):
      for j in range(n):
        x = i * n + j
        output[x // c][x % c] = mat[i][j]

    return output
```

```python
mat = [[1, 2], [3, 4]]
r = 1
c = 4
Solution().matrixReshape(mat=mat, r=r, c=c)
```
