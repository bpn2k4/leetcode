```python
from typing import List


class Solution:
  def leftRightDifference(self, nums: List[int]) -> List[int]:
    n = len(nums)
    left_sum = [0] * n
    for i in range(1, n):
      left_sum[i] = left_sum[i - 1] + nums[i - 1]
    total = sum(nums)
    output = []
    for i in range(n):
      right_sum = total - left_sum[i] - nums[i]
      output.append(abs(left_sum[i] - right_sum))
    return output
```

```python
nums = [10, 4, 8, 3]
Solution().leftRightDifference(nums=nums)
```

```python
nums = [1]
Solution().leftRightDifference(nums=nums)
```
