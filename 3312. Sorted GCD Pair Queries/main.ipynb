```python
from typing import List
import bisect


class Solution:
  def gcdValues(self, nums: List[int], queries: List[int]) -> List[int]:
    m = max(nums)
    cnt = [0] * (m + 1)
    for num in nums:
      cnt[num] += 1
    for i in range(1, m + 1):
      for j in range(i * 2, m + 1, i):
        cnt[i] += cnt[j]
    for i in range(1, m + 1):
      cnt[i] = cnt[i] * (cnt[i] - 1) // 2
    for i in range(m, 0, -1):
      for j in range(i * 2, m + 1, i):
        cnt[i] -= cnt[j]
    for i in range(1, m + 1):
      cnt[i] += cnt[i - 1]
    ans = []
    for q in queries:
      q += 1
      pos = bisect.bisect_left(cnt, q)
      ans.append(pos)
    return ans
```

```python
nums = [2, 3, 4]
queries = [0, 2, 2]
Solution().gcdValues(nums=nums, queries=queries)
```

```python
nums = [4, 4, 2, 1]
queries = [5, 3, 1, 0]
Solution().gcdValues(nums=nums, queries=queries)
```

```python
nums = [2, 2]
queries = [0, 0]
Solution().gcdValues(nums=nums, queries=queries)
```
