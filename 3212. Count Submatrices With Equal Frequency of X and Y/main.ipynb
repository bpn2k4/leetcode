```python
from typing import List


class Solution:
  def numberOfSubmatrices(self, grid: List[List[str]]) -> int:
    m, n = len(grid), len(grid[0])
    count_x = [[0] * (n + 1) for _ in range(m + 1)]
    count_y = [[0] * (n + 1) for _ in range(m + 1)]
    count = 0
    for i in range(1, m + 1):
      for j in range(1, n + 1):
        count_x[i][j] = (
            count_x[i - 1][j]
            + count_x[i][j - 1]
            - count_x[i - 1][j - 1]
            + (grid[i - 1][j - 1] == 'X')
        )
        count_y[i][j] = (
            count_y[i - 1][j]
            + count_y[i][j - 1]
            - count_y[i - 1][j - 1]
            + (grid[i - 1][j - 1] == 'Y')
        )
        if count_x[i][j] > 0 and count_x[i][j] == count_y[i][j]:
          count += 1
    return count
```

```python
grid = [["X", "Y", "."], ["Y", ".", "."]]
Solution().numberOfSubmatrices(grid=grid)
```
