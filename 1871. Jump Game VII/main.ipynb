```python
class Solution:
  def canReach(self, s: str, minJump: int, maxJump: int) -> bool:
    n = len(s)
    dp = [0] * n
    prefix = [0] * n
    dp[0] = 1
    for i in range(minJump):
      prefix[i] = 1
    for i in range(minJump, n):
      left, right = i - maxJump, i - minJump
      if s[i] == '0':
        if left < 0:
          total = prefix[right]
        else:
          total = prefix[right] - prefix[left - 1]
        if total == 0:
          dp[i] == 0
        else:
          dp[i] = 1
      prefix[i] = prefix[i - 1] + dp[i]
    return bool(dp[n - 1])
```

```python
s = "011010"
minJump = 2
maxJump = 3
Solution().canReach(s=s, minJump=minJump, maxJump=maxJump)
```
