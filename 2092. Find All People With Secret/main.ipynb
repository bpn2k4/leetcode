```python
from typing import List
import heapq
from collections import defaultdict


class Solution:
  def findAllPeople(self, n: int, meetings: List[List[int]], firstPerson: int) -> List[int]:
    graph = defaultdict(list)
    for x, y, t in meetings:
      graph[x].append((t, y))
      graph[y].append((t, x))

    pq = []
    heapq.heappush(pq, (0, 0))
    heapq.heappush(pq, (0, firstPerson))
    visited = [False] * n

    while pq:
      time, person = heapq.heappop(pq)
      if visited[person]:
        continue
      visited[person] = True
      for t, next_person in graph[person]:
        if not visited[next_person] and t >= time:
          heapq.heappush(pq, (t, next_person))

    return [i for i in range(n) if visited[i]]
```

```python
n = 6
meetings = [[1, 2, 5], [2, 3, 8], [1, 5, 10]]
firstPerson = 1
Solution().findAllPeople(n=n, meetings=meetings, firstPerson=firstPerson)
```
