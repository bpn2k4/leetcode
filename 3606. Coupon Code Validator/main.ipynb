```python
from typing import List


class Solution:
  def validateCoupons(self, code: List[str], businessLine: List[str], isActive: List[bool]) -> List[str]:
    
    business_line2priority = {
      'electronics': 0,
      'grocery': 1,
      'pharmacy': 2,
      'restaurant': 3
    }
    n = len(code)
    valid_codes = []
    for i in range(n):
      if isActive[i] and businessLine[i] in business_line2priority and code[i].replace('_', 'a').isalnum():
        valid_codes.append((code[i], business_line2priority[businessLine[i]]))
    valid_codes.sort(key=lambda x: (x[1], x[0]))
    output = [x[0] for x in valid_codes]
    return output
```

```python
code = ["SAVE20", "", "PHARMA5", "SAVE@20"]
businessLine = ["restaurant", "grocery", "pharmacy", "restaurant"]
isActive = [True, True, True, True]
Solution().validateCoupons(code=code, businessLine=businessLine, isActive=isActive)
```
