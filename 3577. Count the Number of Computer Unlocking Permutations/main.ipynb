```python
from typing import List


class Solution:
  def countPermutations(self, complexity: List[int]) -> int:
    MOD = 10**9 + 7
    n = len(complexity)
    for i in range(1, n):
      if complexity[i] <= complexity[0]:
        return 0
      
    count = 1
    for i in range(2, n):
      count = count * i % MOD

    return count
```

```python
complexity = [1, 2, 3]
Solution().countPermutations(complexity=complexity)
```

```python
complexity = [3, 3, 3, 4, 4, 4]
Solution().countPermutations(complexity=complexity)
```
