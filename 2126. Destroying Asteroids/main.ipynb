```python
from typing import List


class Solution:
  def asteroidsDestroyed(self, mass: int, asteroids: List[int]) -> bool:
    asteroids.sort()
    for i in range(len(asteroids) - 1):
      if mass < asteroids[i]:
        return False
      mass += asteroids[i]
    if mass < asteroids[-1]:
      return False
    return True
```

```python
mass = 10
asteroids = [3, 9, 19, 5, 21]
Solution().asteroidsDestroyed(mass=mass, asteroids=asteroids)
```
