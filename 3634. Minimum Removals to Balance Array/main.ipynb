```python
from typing import List


class Solution:
  def minRemoval(self, nums: List[int], k: int) -> int:
    n = len(nums)
    nums.sort()
    result = n
    right = 0
    for left in range(n):
      while right < n and nums[right] <= nums[left] * k:
        right += 1
      result = min(result, n - (right - left))
    return result
```

```python
nums = [2, 1, 5]
k = 2
Solution().minRemoval(nums=nums, k=k)
```

```python
nums = [1, 6, 2, 9]
k = 3
Solution().minRemoval(nums=nums, k=k)
```
