```python
from typing import List


class Solution:
  def containsCycle(self, grid: List[List[str]]) -> bool:
    m, n = len(grid), len(grid[0])
    visited = [[False] * n for _ in range(m)]
    
    def dfs(r: int, c: int, pr: int, pc: int) -> bool:
      visited[r][c] = True
      for dx, dy in [(0, 1), (0, -1), (1, 0), (-1, 0)]:
        nr, nc = r + dx, c + dy
        if (nr, nc) != (pr, pc):
          if 0 <= nr < m and 0 <= nc < n:
            if grid[nr][nc] == grid[r][c]:
              if visited[nr][nc]:
                return True
              if dfs(nr, nc, r, c):
                return True
      return False

    for i in range(m):
      for j in range(n):
        if not visited[i][j]:
          if dfs(i, j, -1, -1):
            return True
    return False
```

```python
grid = [["d","b","b"],["c","a","a"],["b","a","c"],["c","c","c"],["d","d","a"]]
for i in grid:
  print(i)
Solution().containsCycle(grid=grid)
```
