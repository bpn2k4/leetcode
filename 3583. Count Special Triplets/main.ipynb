```python
from typing import List
from collections import Counter


class Solution:
  def specialTriplets(self, nums: List[int]) -> int:
    
    MOD = 10**9 + 7
    num2count = Counter(nums)
    count = 0
    num2fred = {}
    for num in nums:
      num2count[num] -= 1
      if num * 2 in num2fred:
        count_left = num2fred[num * 2]
        count_right = num2count[num * 2]
        if count_left > 0 and count_right > 0:
          count = (count + count_left * count_right) % MOD
      num2fred[num] = num2fred.get(num, 0) + 1

    return count
```

```python
nums = [6, 3, 6]
Solution().specialTriplets(nums=nums)
```

```python
nums = [0, 1, 0, 0]
Solution().specialTriplets(nums=nums)
```

```python
nums = [8, 4, 2, 8, 4]
Solution().specialTriplets(nums=nums)
```
