```python
from typing import List


class Solution:
  def countCoveredBuildings(self, n: int, buildings: List[List[int]]) -> int:
    x_min = [n + 1] * (n + 1)
    x_max = [0] * (n + 1)
    y_min = [n + 1] * (n + 1)
    y_max = [0] * (n + 1)
    for x, y in buildings:
      y_min[x] = min(y_min[x], y)
      y_max[x] = max(y_max[x], y)
      x_min[y] = min(x_min[y], x)
      x_max[y] = max(x_max[y], x)

    count = 0
    for x, y in buildings:
      if x_min[y] < x < x_max[y] and y_min[x] < y < y_max[x]:
        count += 1
    
    return count
```

```python
n = 3
buildings = [[1, 2], [2, 2], [3, 2], [2, 1], [2, 3]]
Solution().countCoveredBuildings(n=n, buildings=buildings)
```
