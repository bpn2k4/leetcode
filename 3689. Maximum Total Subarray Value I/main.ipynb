```python
from typing import List


class Solution:
  def maxTotalValue(self, nums: List[int], k: int) -> int:
    return (max(nums) - min(nums)) * k
```

```python
nums = [1, 3, 2]
k = 2
Solution().maxTotalValue(nums=nums, k=k)
```

```python
nums = [4, 2, 5, 1]
k = 3
Solution().maxTotalValue(nums=nums, k=k)
```
