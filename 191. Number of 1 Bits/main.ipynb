```python
class Solution:
  prefix = [1] * 32
  for i in range(31, 0, -1):
    prefix[i - 1] = prefix[i] * 2

  def hammingWeight(self, n: int) -> int:
    count = 0
    for num in self.prefix:
      if n >= num:
        count += 1
        n -= num
    return count
```

```python
n = 11
Solution().hammingWeight(n=n)
```

```python
n = 128
Solution().hammingWeight(n=n)
```

```python
n = 2147483645
Solution().hammingWeight(n=n)
```
