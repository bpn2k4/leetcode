```python
class Solution:
  def numberOfStableArrays(self, zero: int, one: int, limit: int) -> int:
    MOD = 10**9 + 7
    dp = [[[0, 0] for _ in range(one + 1)] for _ in range(zero + 1)]
    for i in range(zero + 1):
      for j in range(one + 1):
        for lastBit in range(2):
          if i == 0:
            if lastBit == 0 or j > limit:
              dp[i][j][lastBit] = 0
            else:
              dp[i][j][lastBit] = 1
          elif j == 0:
            if lastBit == 1 or i > limit:
              dp[i][j][lastBit] = 0
            else:
              dp[i][j][lastBit] = 1
          elif lastBit == 0:
            dp[i][j][lastBit] = dp[i - 1][j][0] + dp[i - 1][j][1]
            if i > limit:
              dp[i][j][lastBit] -= dp[i - limit - 1][j][1]
          else:
            dp[i][j][lastBit] = dp[i][j - 1][0] + dp[i][j - 1][1]
            if j > limit:
              dp[i][j][lastBit] -= dp[i][j - limit - 1][0]
          dp[i][j][lastBit] %= MOD
    return (dp[-1][-1][0] + dp[-1][-1][1]) % MOD
```

```python
zero = 1
one = 1
limit = 2
Solution().numberOfStableArrays(zero=zero, one=one, limit=limit)
```
