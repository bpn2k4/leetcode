```python
from typing import List


class Solution:
  def uniqueXorTriplets(self, nums: List[int]) -> int:
    max_num = max(nums)
    u = 1
    while u <= max_num:
      u <<= 1
    one = [False] * u
    two = [False] * u
    three = [False] * u
    for num in nums:
      one[num] = True
      for x in range(u):
        if one[x]:
          two[x ^ num] = True
    for num in nums:
      for x in range(u):
        if two[x]:
          three[x ^ num] = True
    return sum(1 for b in three if b)
```

```python
nums = [1, 3]
Solution().uniqueXorTriplets(nums=nums)
```

```python
nums = [6, 7, 8, 9]
Solution().uniqueXorTriplets(nums=nums)
```
