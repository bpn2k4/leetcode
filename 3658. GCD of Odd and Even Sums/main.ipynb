```python
import math


class Solution:
  def gcdOfOddEvenSums(self, n: int) -> int:
    odd_sum = 0
    even_sum = 0
    for i in range(n):
      odd_sum += 2 * i + 1
      even_sum += 2 * (i + 1)
    return math.gcd(odd_sum, even_sum)
```

```python
n = 4
Solution().gcdOfOddEvenSums(n=n)
```

```python
n = 5
Solution().gcdOfOddEvenSums(n=n)
```
