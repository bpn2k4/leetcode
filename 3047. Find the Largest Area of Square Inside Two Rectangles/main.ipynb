```python
from typing import List


class Solution:
  def largestSquareArea(self, bottomLeft: List[List[int]], topRight: List[List[int]]) -> int:
    n = len(bottomLeft)
    max_side = 0

    for i in range(n):
      a1, b1 = bottomLeft[i]
      c1, d1 = topRight[i]

      for j in range(i + 1, n):
        a2, b2 = bottomLeft[j]
        c2, d2 = topRight[j]

        overlap_x = min(c1, c2) - max(a1, a2)
        overlap_y = min(d1, d2) - max(b1, b2)

        if overlap_x > 0 and overlap_y > 0:
          max_side = max(max_side, min(overlap_x, overlap_y))

    return max_side * max_side
```

```python
bottomLeft = [[1, 1], [2, 2], [3, 1]]
topRight = [[3, 3], [4, 4], [6, 6]]
Solution().largestSquareArea(bottomLeft=bottomLeft, topRight=topRight)
```
