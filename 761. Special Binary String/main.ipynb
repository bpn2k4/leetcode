```python
class Solution:
  def makeLargestSpecial(self, s: str) -> str:
    count = 0
    i = 0
    res = []
    for j in range(len(s)):
      count += 1 if s[j] == '1' else -1
      if count == 0:
        res.append('1' + self.makeLargestSpecial(s[i + 1:j]) + '0')
        i = j + 1
    res.sort(reverse=True)
    return ''.join(res)
```

```python
s = "11011000"
Solution().makeLargestSpecial(s=s)
```

```python
s = "10"
Solution().makeLargestSpecial(s=s)
```
