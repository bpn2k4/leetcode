```python
from typing import List


class Solution:
  def minimumEffort(self, tasks: List[List[int]]) -> int:
    tasks.sort(key=lambda x: x[1] - x[0])
    ans = 0
    for task in tasks:
      ans = max(ans + task[0], task[1])
    return ans
```

```python
tasks = [[1, 2], [2, 4], [4, 8]]
Solution().minimumEffort(tasks=tasks)
```

```python
tasks = [[1, 3], [2, 4], [10, 11], [10, 12], [8, 9]]
Solution().minimumEffort(tasks=tasks)
```
