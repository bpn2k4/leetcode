```python
class Solution:
  def longestBalanced(self, s: str) -> int:
    prefix = [(0, 0, 0)]
    for char in s:
      a, b, c = prefix[-1]
      if char == 'a':
        a += 1
      elif char == 'b':
        b += 1
      else:
        c += 1
      prefix.append((a, b, c))
    max_length = 0
    first_seen = {}
    for i, (a, b, c) in enumerate(prefix):
      keys = [
        ('abc', a - b, a - c),
        ('ab', a - b, c),
        ('bc', b - c, a),
        ('ca', c - a, b),
        ('a', b, c),
        ('b', a, c),
        ('c', a, b),
      ]
      for key in keys:
        if key in first_seen:
          max_length = max(max_length, i - first_seen[key])
        else:
          first_seen[key] = i
    return max_length
```

```python
s = "abbac"
Solution().longestBalanced(s=s)
```

```python
s = "aabcc"
Solution().longestBalanced(s=s)
```
