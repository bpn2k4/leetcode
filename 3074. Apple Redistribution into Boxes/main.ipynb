```python
from typing import List


class Solution:
  def minimumBoxes(self, apple: List[int], capacity: List[int]) -> int:
    total_apple = sum(apple)
    capacity.sort(reverse=True)
    sum_capacity = 0
    for i in range(len(capacity)):
      sum_capacity += capacity[i]
      if sum_capacity >= total_apple:
        return i + 1
    return -1
```

```python
apple = [1, 3, 2]
capacity = [4, 3, 1, 5, 2]
Solution().minimumBoxes(apple=apple, capacity=capacity)
```

```python
apple = [5, 5, 5]
capacity = [2, 4, 2, 7]
Solution().minimumBoxes(apple=apple, capacity=capacity)
```
