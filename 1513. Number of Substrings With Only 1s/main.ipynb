```python
class Solution:
  def numSub(self, s: str) -> int:
    
    MOD = 10**9 + 7
    count = 0
    s += '0'
    count_one = 0
    for char in s:
      if char == '1':
        count_one += 1
      else:
        count += count_one * (count_one + 1) // 2
        count %= MOD
        count_one = 0

    return count
```

```python
s = "0110111"
Solution().numSub(s=s)
```

```python
s = "101"
Solution().numSub(s=s)
```

```python
s = "111111"
Solution().numSub(s=s)
```
