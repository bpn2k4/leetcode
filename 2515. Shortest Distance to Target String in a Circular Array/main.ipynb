```python
from typing import List


class Solution:
  def closestTarget(self, words: List[str], target: str, startIndex: int) -> int:
    n = len(words)
    min_distance = n
    for i in range(n):
      if words[(startIndex + i) % n] == target:
        min_distance = min(min_distance, i)
        break
    for i in range(n):
      if words[(startIndex - i) % n] == target:
        min_distance = min(min_distance, i)
        break
    return -1 if min_distance == n else min_distance
```

```python
words = ["hello", "i", "am", "leetcode", "hello"]
target = "hello"
startIndex = 1
Solution().closestTarget(words=words, target=target, startIndex=startIndex)
```
