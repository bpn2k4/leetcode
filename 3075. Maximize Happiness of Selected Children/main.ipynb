```python
from typing import List


class Solution:
  def maximumHappinessSum(self, happiness: List[int], k: int) -> int:
    happiness.sort(reverse=True)
    sum_happiness = 0
    for i in range(k):
      sum_happiness += max(0, happiness[i] - i)  
    return sum_happiness
```

```python
happiness = [1, 2, 3]
k = 2
Solution().maximumHappinessSum(happiness=happiness, k=k)
```

```python
happiness = [1, 1, 1, 1]
k = 2
Solution().maximumHappinessSum(happiness=happiness, k=k)
```

```python
happiness = [2, 3, 4, 5]
k = 1
Solution().maximumHappinessSum(happiness=happiness, k=k)
```
