```python
from typing import List


class Solution:
  def numberOfPaths(self, grid: List[List[int]], k: int) -> int:
    
    MOD = 10**9 + 7
    m, n = len(grid), len(grid[0])
    dp = [[[0 for _ in range(k)] for _ in range(n + 1)] for _ in range(m + 1)]
    for i in range(1, m + 1):
      for j in range(1, n + 1):
        if i == 1 and j == 1:
          dp[i][j][grid[i - 1][j - 1] % k] = 1
        else:
          v = grid[i - 1][j - 1] % k
          for r in range(k):
            p = (r - v + k) % k
            dp[i][j][r] = (dp[i - 1][j][p] + dp[i][j - 1][p]) % MOD
          
    return dp[m][n][0]
```

```python
grid = [[5, 2, 4], [3, 0, 5], [0, 7, 2]]
k = 3
Solution().numberOfPaths(grid=grid, k=k)
```

```python
grid = [[0, 0]]
k = 5
Solution().numberOfPaths(grid=grid, k=k)
```
