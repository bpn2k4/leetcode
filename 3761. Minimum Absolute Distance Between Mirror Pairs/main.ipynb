```python
from typing import List


class Solution:
  def minMirrorPairDistance(self, nums: List[int]) -> int:
    mark = {}
    n = len(nums)
    min_distance = n + 1
    for i in range(n):
      num = nums[i]
      if num in mark:
        min_distance = min(min_distance, i - mark[num])
      reverse = int(str(num)[::-1])
      mark[reverse] = i
    return min_distance if min_distance != n + 1 else -1
```

```python
nums = [12, 21, 45, 33, 54]
Solution().minMirrorPairDistance(nums=nums)
```

```python
nums = [120, 21]
Solution().minMirrorPairDistance(nums=nums)
```
