```python
class Solution:
  def longestBalanced(self, s: str) -> int:
    n = len(s)
    max_length = 0
    for i in range(n):
      char2count = {}
      for j in range(i, n):
        char2count[s[j]] = char2count.get(s[j], 0) + 1
        if len(set(char2count.values())) == 1:
          max_length = max(max_length, j - i + 1)
    return max_length
```

```python
s = "abbac"
Solution().longestBalanced(s=s)
```

```python
s = "zzabccy"
Solution().longestBalanced(s=s)
```
