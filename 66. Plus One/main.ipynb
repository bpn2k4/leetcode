```python
from typing import List


class Solution:
  def plusOne(self, digits: List[int]) -> List[int]:
    r = 1
    for i in range(len(digits) - 1, -1, -1):
      num = digits[i] + r
      digits[i] = num % 10
      r = num // 10
    if r:
      digits = [1] + digits
    return digits
```

```python
digits = [1, 2, 3]
Solution().plusOne(digits=digits)
```

```python
digits = [4, 3, 2, 1]
Solution().plusOne(digits=digits)
```

```python
digits = [9]
Solution().plusOne(digits=digits)
```
