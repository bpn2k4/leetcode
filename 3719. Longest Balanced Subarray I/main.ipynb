```python
from typing import List


class Solution:
  def longestBalanced(self, nums: List[int]) -> int:
    n = len(nums)
    max_length = 0
    for i in range(n):
      odd = {}
      even = {}
      for j in range(i, n):
        if nums[j] % 2 == 0:
          even[nums[j]] = even.get(nums[j], 0) + 1
        else:
          odd[nums[j]] = odd.get(nums[j], 0) + 1
        if len(odd) == len(even):
          max_length = max(max_length, j - i + 1)
    return max_length
```

```python
nums = [2, 5, 4, 3]
Solution().longestBalanced(nums=nums)
```

```python
nums = [3, 2, 2, 5, 4]
Solution().longestBalanced(nums=nums)
```
