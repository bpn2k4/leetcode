```python
from typing import List


class Solution:
  def earliestFinishTime(
    self,
    landStartTime: List[int],
    landDuration: List[int],
    waterStartTime: List[int],
    waterDuration: List[int],
  ) -> int:

    def solve(start1: List[int], duration1: List[int], start2: List[int], duration2: List[int]):
      finish1 = float('inf')
      for i in range(len(start1)):
        finish1 = min(finish1, start1[i] + duration1[i])
      finish2 = float('inf')
      for i in range(len(start2)):
        finish2 = min(finish2, max(start2[i], finish1) + duration2[i])
      return finish2

    land_water = solve(
      start1=landStartTime, 
      duration1=landDuration, 
      start2=waterStartTime, 
      duration2=waterDuration,
    )
    water_land = solve(
      start1=waterStartTime, 
      duration1=waterDuration,
      start2=landStartTime,
      duration2=landDuration,
    )
    return min(land_water, water_land)
```

```python
landStartTime = [2, 8]
landDuration = [4, 1]
waterStartTime = [6]
waterDuration = [3]
Solution().earliestFinishTime(
  landStartTime=landStartTime,
  landDuration=landDuration,
  waterStartTime=waterStartTime,
  waterDuration=waterDuration,
)
```
