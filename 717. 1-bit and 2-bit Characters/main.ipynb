```python
from typing import List


class Solution:
  def isOneBitCharacter(self, bits: List[int]) -> bool:
    
    n = len(bits)
    i = 0
    while i < len(bits):
      if bits[i] == 1:
        i += 2
        continue
      if bits[i] == 0:
        if i == n - 1:
          return True
        else:
          i += 1

    return False
```

```python
bits = [1, 0, 0]
Solution().isOneBitCharacter(bits=bits)
```

```python
bits = [1, 1, 1, 0]
Solution().isOneBitCharacter(bits=bits)
```
