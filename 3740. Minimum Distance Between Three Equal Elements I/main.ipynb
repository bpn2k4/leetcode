```python
from typing import List


class Solution:
  def minimumDistance(self, nums: List[int]) -> int:
    num2indies = {}
    for index, num in enumerate(nums):
      if num not in num2indies:
        num2indies[num] = []
      num2indies[num].append(index)
    min_distance = 100000
    for indies in num2indies.values():
      for i in range(len(indies) - 2):
        distance = indies[i + 1] - indies[i] + indies[i + 2] - indies[i + 1] + indies[i + 2] - indies[i]
        min_distance = min(min_distance, distance)
    return min_distance if min_distance != 100000 else -1
```

```python
nums = [1, 2, 1, 1, 3]
Solution().minimumDistance(nums=nums)
```

```python
nums = [1, 1, 2, 3, 2, 1, 2]
Solution().minimumDistance(nums=nums)
```
