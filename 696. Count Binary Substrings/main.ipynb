```python
class Solution:
  def countBinarySubstrings(self, s: str) -> int:
    consecutive_groups = [1]
    for i in range(1, len(s)):
      if s[i] == s[i - 1]:
        consecutive_groups[-1] += 1
      else:
        consecutive_groups.append(1)
    count = 0
    for i in range(1, len(consecutive_groups)):
      count += min(consecutive_groups[i], consecutive_groups[i - 1])
    return count
```

```python
s = "00110011"
Solution().countBinarySubstrings(s=s)
```

```python
s = "10101"
Solution().countBinarySubstrings(s=s)
```
