```python
from typing import List


class Solution:
  def numSpecial(self, mat: List[List[int]]) -> int:
    m, n = len(mat), len(mat[0])
    count_row = [0] * m
    count_col = [0] * n
    for i in range(m):
      count_row[i] = sum(mat[i])
    for j in range(n):
      count_col[j] = sum(mat[k][j] for k in range(m))
    total_special_points = 0
    for i in range(m):
      for j in range(n):
        if mat[i][j] == 1 and count_row[i] == 1 and count_col[j] == 1:
          total_special_points += 1
    return total_special_points
```

```python
mat = [[1, 0, 0], [0, 0, 1], [1, 0, 0]]
Solution().numSpecial(mat=mat)
```
