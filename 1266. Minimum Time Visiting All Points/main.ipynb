```python
from typing import List


class Solution:
  def minTimeToVisitAllPoints(self, points: List[List[int]]) -> int:
    time = 0
    for i in range(1, len(points)):
      x1, y1 = points[i - 1]
      x2, y2 = points[i]
      dx = abs(x1 - x2)
      dy = abs(y1 - y2)
      if dx == dy:
        time += dx
      else:
        time += min(dx, dy) + abs(dx - dy)
    return time
```

```python
points = [[1, 1], [3, 4], [-1, 0]]
Solution().minTimeToVisitAllPoints(points=points)
```

```python
points = [[3, 2], [-2, 2]]
Solution().minTimeToVisitAllPoints(points=points)
```
