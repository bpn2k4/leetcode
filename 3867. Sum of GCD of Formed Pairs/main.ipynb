```python
import math


class Solution:
  def gcdSum(self, nums: list[int]) -> int:
    n = len(nums)
    mx = []
    prefixMax = - math.inf
    for x in nums:
      prefixMax = max(prefixMax, x)
      mx.append(prefixMax)
    prefix_gcd = [math.gcd(x, y) for x, y in zip(nums, mx)]
    prefix_gcd.sort()
    ans = 0
    left, right = 0, n - 1
    while left < right:
      ans += math.gcd(prefix_gcd[left], prefix_gcd[right])
      left += 1
      right -= 1
    return ans
```

```python
nums = [2, 6, 4]
Solution().gcdSum(nums=nums)
```

```python
nums = [3, 6, 2, 8]
Solution().gcdSum(nums=nums)
```
