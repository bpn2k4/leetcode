```python
from typing import List


class Solution:
  def minimumCost(self, nums: List[int]) -> int:
    min_cost = 151
    n = len(nums)
    for i in range(1, n):
      for j in range(i + 1, n):
        min_cost = min(min_cost, nums[i] + nums[j])
    return min_cost + nums[0]
```

```python
nums = [1, 2, 3, 12]
Solution().minimumCost(nums=nums)
```

```python
nums = [5, 4, 3]
Solution().minimumCost(nums=nums)
```

```python
nums = [10, 3, 1, 1]
Solution().minimumCost(nums=nums)
```
