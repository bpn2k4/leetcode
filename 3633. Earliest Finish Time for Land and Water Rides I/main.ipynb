```python
from typing import List


class Solution:
  def earliestFinishTime(
    self,
    landStartTime: List[int],
    landDuration: List[int],
    waterStartTime: List[int],
    waterDuration: List[int],
  ) -> int:
    earliest_finish_time = 9999999999
    for i in range(len(landStartTime)):
      for j in range(len(waterStartTime)):
        land_start, land_end = landStartTime[i], landStartTime[i] + landDuration[i]
        water_start, water_end = waterStartTime[j], waterStartTime[j] + waterDuration[j]
        if water_start < land_end:
          earliest_finish_time = min(earliest_finish_time, land_end + waterDuration[j])
        else:
          earliest_finish_time = min(earliest_finish_time, water_end)
        if land_start < water_end:
          earliest_finish_time = min(earliest_finish_time, landDuration[i] + water_end)
        else:
          earliest_finish_time = min(earliest_finish_time, land_end)
    return earliest_finish_time
```

```python
landStartTime = [2, 8]
landDuration = [4, 1]
waterStartTime = [6]
waterDuration = [3]
Solution().earliestFinishTime(
  landStartTime=landStartTime,
  landDuration=landDuration,
  waterStartTime=waterStartTime,
  waterDuration=waterDuration,
)
```
