```python
from typing import List
from collections import Counter


class Solution:
  def getSneakyNumbers(self, nums: List[int]) -> List[int]:
    
    num2count = Counter(nums)
    
    return [num for num, count in num2count.items() if count > 1]
```

```python
nums = [0, 1, 1, 0]
Solution().getSneakyNumbers(nums=nums)
```

```python
nums = [0, 3, 2, 1, 3, 2]
Solution().getSneakyNumbers(nums=nums)
```

```python
nums = [7, 1, 5, 4, 3, 4, 6, 0, 9, 5, 8, 2]
Solution().getSneakyNumbers(nums=nums)
```
