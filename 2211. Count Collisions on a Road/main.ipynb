```python
class Solution:
  def countCollisions(self, directions: str) -> int:
    
    RIGHT = 'R'
    STOP = 'S'
    LEFT = 'L'
    count = 0
    r = 0
    last = ''
    for direction in directions:
      if last == '':
        last = direction
        if direction == RIGHT:
          r += 1
        continue
      if direction == STOP:
        count += r
        r = 0
        last = STOP
        continue
      if direction == RIGHT:
        r += 1
        last = RIGHT
        continue
      if direction == LEFT:
        if last == RIGHT:
          count += r + 1
          r = 0
          last = STOP
          continue
        if last == STOP:
          count += 1
          last = STOP
          continue
                
    return count
```

```python
directions = "LLRLRLLSLRLLSLSSSS"
Solution().countCollisions(directions=directions)
```
