```python
from typing import List


class Solution:
  def xorAfterQueries(self, nums: List[int], queries: List[List[int]]) -> int:
    MOD = 10**9 + 7
    for l, r, k, v in queries:
      for i in range(l, r + 1, k):
        nums[i] = nums[i] * v % MOD
    output = 0
    for num in nums:
      output ^= num
    return output
```

```python
nums = [1, 1, 1]
queries = [[0, 2, 1, 4]]
Solution().xorAfterQueries(nums=nums, queries=queries)
```

```python
nums = [2, 3, 1, 5, 4]
queries = [[1, 4, 2, 3], [0, 2, 1, 2]]
Solution().xorAfterQueries(nums=nums, queries=queries)
```
