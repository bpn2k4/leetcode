```python
from typing import List


class Solution:
  def numOfStrings(self, patterns: List[str], word: str) -> int:
    return sum(1 for i in patterns if i in word)
```

```python
patterns = ["a", "abc", "bc", "d"]
word = "abc"
Solution().numOfStrings(patterns=patterns, word=word)
```

```python
patterns = ["a", "b", "c"]
word = "aaaaabbbbb"
Solution().numOfStrings(patterns=patterns, word=word)
```

```python
patterns = ["a", "a", "a"]
word = "ab"
Solution().numOfStrings(patterns=patterns, word=word)
```
