```python
from typing import List
import heapq


class Solution:
  def minimumCost(self, nums: List[int], k: int, dist: int) -> int:
    heap_used, heap_unused, used = [], [], set()
    s, ans = 0, float('inf')
    for right in range(1, len(nums)):
      left = right - dist - 1
      if left > 0 and left in used:
        used.remove(left)
        s -= nums[left]
        while heap_unused and heap_unused[0][1] < left:
          heapq.heappop(heap_unused)
        if heap_unused:
          num, i = heapq.heappop(heap_unused)
          heapq.heappush(heap_used, (-num, i))
          used.add(i)
          s += num
      if len(used) < k - 1:
        heapq.heappush(heap_used, (-nums[right], right))
        used.add(right)
        s += nums[right]
      else:
        while heap_used[0][1] not in used:
          heapq.heappop(heap_used)
        if nums[right] < -heap_used[0][0]:
          num, i = heapq.heapreplace(heap_used, (-nums[right], right))
          used.remove(i)
          used.add(right)
          s += num + nums[right]
          heapq.heappush(heap_unused, (-num, i))
        else:
          heapq.heappush(heap_unused, (nums[right], right))
      if left >= 0:
        ans = min(s, ans)
    return nums[0] + ans
```

```python
nums = [1, 3, 2, 6, 4, 2]
k = 3
dist = 3
Solution().minimumCost(nums=nums, k=k, dist=dist)
```
