```python
from typing import List


class Solution:
  def minNumberOfSeconds(self, mountainHeight: int, workerTimes: List[int]) -> int:
    max_worker_time = max(workerTimes)
    left = 1
    right = max_worker_time * mountainHeight * (mountainHeight + 1) // 2
    ans = 0
    epsilon = 1e-6
    while left <= right:
      mid = (left + right) // 2
      count = 0
      for time in workerTimes:
        work = mid // time
        k = int((-1 + ((1 + work * 8) ** 0.5)) / 2 + epsilon)
        count += k
      if count >= mountainHeight:
        ans = mid
        right = mid - 1
      else:
        left = mid + 1
    return ans
```

```python
mountainHeight = 4
workerTimes = [2, 1, 1]
Solution().minNumberOfSeconds(mountainHeight=mountainHeight, workerTimes=workerTimes)
```
