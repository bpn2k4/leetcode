```python
from typing import List


class Solution:
  def minimumOperations(self, nums: List[int]) -> int:
    
    return len(nums) - len([num for num in nums if num % 3 == 0])
```

```python
nums = [1, 2, 3, 4]
Solution().minimumOperations(nums=nums)
```

```python
nums = [3, 6, 9]
Solution().minimumOperations(nums=nums)
```
