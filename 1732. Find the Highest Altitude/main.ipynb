```python
from typing import List


class Solution:
  def largestAltitude(self, gain: List[int]) -> int:
    height = 0
    max_height = 0
    for g in gain:
      height += g
      max_height = max(max_height, height)
    return max_height
```

```python
gain = [-5, 1, 5, 0, -7]
Solution().largestAltitude(gain=gain)
```

```python
gain = [-4, -3, -2, -1, 4, 3, 2]
Solution().largestAltitude(gain=gain)
```
