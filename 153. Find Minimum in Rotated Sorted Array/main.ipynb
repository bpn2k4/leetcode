```python
from typing import List


class Solution:
  def findMin(self, nums: List[int]) -> int:
    l, r = 0, len(nums) - 1
    while l < r:
      mid = (l + r) // 2
      if nums[mid] > nums[-1]:
        l = mid + 1
      else:
        r = mid
    return nums[l]
```

```python
nums = [3, 4, 5, 1, 2]
Solution().findMin(nums=nums)
```

```python
nums = [4, 5, 6, 7, 0, 1, 2]
Solution().findMin(nums=nums)
```

```python
nums = [11, 13, 15, 17]
Solution().findMin(nums=nums)
```
