```python
class Solution:
  def countTriples(self, n: int) -> int:
    n2 = n * n
    square =[i * i for i in range(n + 1)]
    mark_square = {i: 1 for i in square}
    count = 0
    for a in range(1, n + 1):
      for b in range(a + 1, n + 1):
        c = square[a] + square[b]
        if c > n2:
          break
        if c in mark_square:
          count += 1
    return count * 2
```

```python
n = 5
Solution().countTriples(n=n)
```

```python
n = 10
Solution().countTriples(n=n)
```
