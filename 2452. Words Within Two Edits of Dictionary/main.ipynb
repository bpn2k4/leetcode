```python
from typing import List


class Solution:
  def twoEditWords(self, queries: List[str], dictionary: List[str]) -> List[str]:
    word_length = len(queries[0])
    output = []
    for word in queries:
      for word2 in dictionary:
        edit = 0
        for i in range(word_length):
          if word[i] != word2[i]:
            edit += 1
            if edit > 2:
              break
        if edit <= 2:
          output.append(word)
          break
    return output
```

```python
queries = ["word", "note", "ants", "wood"]
dictionary = ["wood", "joke", "moat"]
Solution().twoEditWords(queries=queries, dictionary=dictionary)
```

```python
queries = ["yes"]
dictionary = ["not"]
Solution().twoEditWords(queries=queries, dictionary=dictionary)
```
