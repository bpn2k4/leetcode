```python
from typing import List


class Solution:
  def findRotation(self, mat: List[List[int]], target: List[List[int]]) -> bool:

    n = len(mat)

    def is_equal(mat: List[List[int]], target: List[List[int]]) -> bool:
      for i in range(n):
        for j in range(n):
          if mat[i][j] != target[i][j]:
            return False
      return True

    for _ in range(4):
      for i in range(n // 2):
        for j in range((n + 1) // 2):
          (
              mat[i][j],
              mat[n - 1 - j][i],
              mat[n - 1 - i][n - 1 - j],
              mat[j][n - 1 - i],
          ) = (
              mat[n - 1 - j][i],
              mat[n - 1 - i][n - 1 - j],
              mat[j][n - 1 - i],
              mat[i][j],
          )
      if is_equal(mat, target):
        return True
    return False
```

```python
mat = [[0, 1], [1, 0]]
target = [[1, 0], [0, 1]]
Solution().findRotation(mat=mat, target=target)
```
