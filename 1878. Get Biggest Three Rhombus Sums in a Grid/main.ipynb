```python
from typing import List


class Solution:
  def getBiggestThree(self, grid: List[List[int]]) -> List[int]:

    output = [0, 0, 0]

    def update_output(value: int) -> None:
      if value > output[0]:
        output[0], output[1], output[2] = value, output[0], output[1]
      elif value != output[0] and value > output[1]:
        output[1], output[2] = value, output[1]
      elif value != output[0] and value != output[1] and value > output[2]:
        output[2] = value

    m, n = len(grid), len(grid[0])
    prefix1 = [[0 for _ in range(n + 2)] for _ in range(m + 1)]
    prefix2 = [[0 for _ in range(n + 2)] for _ in range(m + 1)]
    for i in range(1, m + 1):
      for j in range(1, n + 1):
        prefix1[i][j] = prefix1[i - 1][j - 1] + grid[i - 1][j - 1]
        prefix2[i][j] = prefix2[i - 1][j + 1] + grid[i - 1][j - 1]
    for i in range(m):
      for j in range(n):
        update_output(grid[i][j])
        for k in range(i + 2, m, 2):
          ux, uy = i, j
          dx, dy = k, j
          lx, ly = (i + k) // 2, j - (k - i) // 2
          rx, ry = (i + k) // 2, j + (k - i) // 2
          if ly < 0 or ry >= n:
            break
          total = prefix2[lx + 1][ly + 1] - prefix2[ux][uy + 2] \
              + prefix1[rx + 1][ry + 1] - prefix1[ux][uy] \
              + prefix1[dx + 1][dy + 1] - prefix1[lx][ly] \
              + prefix2[dx + 1][dy + 1] - prefix2[rx][ry + 2] \
              - grid[ux][uy] - grid[dx][dy] - grid[lx][ly] - grid[rx][ry]
          update_output(total)
    return [i for i in output if i != 0]
```

```python
grid = [[3, 4, 5, 1, 3], [3, 3, 4, 2, 3], [20, 30, 200, 40, 10], [1, 5, 5, 4, 1], [4, 3, 2, 2, 5]]
Solution().getBiggestThree(grid=grid)
```
