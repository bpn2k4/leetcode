```python
from typing import List


class Solution:
  def sequentialDigits(self, low: int, high: int) -> List[int]:
    result = []
    for i in range(1, 10):
      num = i
      for j in range(i + 1, 10):
        num = num * 10 + j
        if num >= low and num <= high:
          result.append(num)
    result.sort()
    return result
```

```python
low = 100
high = 300
Solution().sequentialDigits(low=low, high=high)
```

```python
low = 1000
high = 13000
Solution().sequentialDigits(low=low, high=high)
```
