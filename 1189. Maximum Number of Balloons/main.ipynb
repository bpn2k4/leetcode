```python
from collections import Counter


class Solution:
  def maxNumberOfBalloons(self, text: str) -> int:
    char2count_ballon = Counter("balloon")
    char2count = Counter(text)
    return min(char2count.get(char, 0) // char2count_ballon[char] for char in char2count_ballon)
```

```python
text = "nlaebolko"
Solution().maxNumberOfBalloons(text=text)
```

```python
text = "loonbalxballpoon"
Solution().maxNumberOfBalloons(text=text)
```
