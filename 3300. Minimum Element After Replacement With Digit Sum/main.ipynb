```python
from typing import List


class Solution:
  def minElement(self, nums: List[int]) -> int:
    for i in range(len(nums)):
      nums[i] = sum(int(i) for i in str(nums[i]))
    return min(nums)
```

```python
nums = [10, 12, 13, 14]
Solution().minElement(nums=nums)
```

```python
nums = [1, 2, 3, 4]
Solution().minElement(nums=nums)
```

```python
nums = [999, 19, 199]
Solution().minElement(nums=nums)
```
