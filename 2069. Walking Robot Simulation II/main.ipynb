```python
from typing import List


class Robot:
  def __init__(self, width: int, height: int):
    self.w = width - 1
    self.h = height - 1
    self.c = (self.w + self.h) * 2
    self.s = 0
    self.moved = False

  def step(self, num: int) -> None:
    self.moved = True
    self.s = (self.s + num) % self.c

  def getPos(self) -> List[int]:
    if 0 <= self.s <= self.w:
      return [self.s, 0]
    elif self.w < self.s <= self.w + self.h:
      return [self.w, self.s - self.w]
    elif self.w + self.h < self.s <= 2 * self.w + self.h:
      return [self.w - (self.s - self.w - self.h), self.h]
    else:
      return [0, self.h - (self.s - 2 * self.w - self.h)]

  def getDir(self) -> str:
    if self.s == 0:
      if self.moved:
        return 'South'
      else:
        return 'East'
    if 1 <= self.s <= self.w:
      return 'East'
    elif self.w < self.s <= self.w + self.h:
      return 'North'
    elif self.w + self.h < self.s <= 2 * self.w + self.h:
      return 'West'
    else:
      return 'South'
```

```python
robot = Robot(6, 3)
print(robot.step(2))
print(robot.step(2))
print(robot.getPos())
print(robot.getDir())
print(robot.step(2))
print(robot.step(1))
print(robot.step(4))
print(robot.getPos())
print(robot.getDir())
```
