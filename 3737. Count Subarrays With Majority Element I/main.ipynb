```python
from typing import List


class Solution:
  def countMajoritySubarrays(self, nums: List[int], target: int) -> int:
    n = len(nums)
    index2count_target = [0] * n
    index2count_target[0] = 1 if nums[0] == target else 0
    for i in range(1, n):
      if nums[i] == target:
        index2count_target[i] = index2count_target[i - 1] + 1
      else:
        index2count_target[i] = index2count_target[i - 1]
    index2count_target.append(0)
    count = 0
    for i in range(n):
      for j in range(i, n):
        target_count = index2count_target[j] - index2count_target[i - 1]
        if target_count > (j - i + 1) // 2:
          count += 1
    return count
```

```python
nums = [1, 2, 2, 3]
target = 2
Solution().countMajoritySubarrays(nums=nums, target=target)
```
