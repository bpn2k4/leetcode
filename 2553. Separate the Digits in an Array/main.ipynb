```python
from typing import List


class Solution:
  def separateDigits(self, nums: List[int]) -> List[int]:
    output = []
    for num in nums:
      for digit in str(num):
        output.append(int(digit))
    return output
```

```python
nums = [13, 25, 83, 77]
Solution().separateDigits(nums=nums)
```

```python
nums = [7, 1, 3, 9]
Solution().separateDigits(nums=nums)
```
