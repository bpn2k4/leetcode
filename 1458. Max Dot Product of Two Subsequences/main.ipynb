```python
from typing import List


class Solution:
  def maxDotProduct(self, nums1: List[int], nums2: List[int]) -> int:
    m, n = len(nums1), len(nums2)
    dp = [[-1000000] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
      for j in range(1, n + 1):
        p = nums1[i - 1] * nums2[j - 1]
        dp[i][j] = max(
          p,
          dp[i - 1][j - 1] + p,
          dp[i - 1][j],
          dp[i][j - 1]
        )
    return dp[m][n]
```

```python
nums1 = [2, 1, -2, 5]
nums2 = [3, 0, -6]
Solution().maxDotProduct(nums1=nums1, nums2=nums2)
```

```python
nums1 = [3, -2]
nums2 = [2, -6, 7]
Solution().maxDotProduct(nums1=nums1, nums2=nums2)
```

```python
nums1 = [-1, -1]
nums2 = [1, 1]
Solution().maxDotProduct(nums1=nums1, nums2=nums2)
```
