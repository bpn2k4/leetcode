```python
from typing import List
class Solution:
  def countSubmatrices(self, grid: List[List[int]], k: int) -> int:
    m, n = len(grid), len(grid[0])
    columns = [0] * n
    count = 0
    for i in range(m):
      row_sum = 0
      for j in range(n):
        columns[j] += grid[i][j]
        row_sum += columns[j]
        if row_sum <= k:
          count += 1
    return count
```

```python
grid = [[7, 6, 3], [6, 6, 1]]
k = 18
Solution().countSubmatrices(grid=grid, k=k)
```

```python
grid = [[7, 2, 9], [1, 5, 0], [2, 6, 6]]
k = 20
Solution().countSubmatrices(grid=grid, k=k)
```
