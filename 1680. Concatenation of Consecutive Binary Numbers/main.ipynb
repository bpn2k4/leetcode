```python
class Solution:
  def concatenatedBinary(self, n: int) -> int:
    MOD = 10**9 + 7
    result = 0
    bit_length = 0
    for i in range(1, n + 1):
      if (i & (i - 1)) == 0:
        bit_length += 1
      result = ((result << bit_length) | i) % MOD
    return result
```

```python
n = 1
Solution().concatenatedBinary(n=n)
```
