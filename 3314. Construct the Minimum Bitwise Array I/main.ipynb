```python
from typing import List


class Solution:
  def minBitwiseArray(self, nums: List[int]) -> List[int]:
    answer = []
    for num in nums:
      candidate = -1
      for j in range(1, num):
        if (j | j + 1) == num:
          candidate = j
          break
      answer.append(candidate)
    return answer
```

```python
nums = [2, 3, 5, 7]
Solution().minBitwiseArray(nums=nums)
```

```python
nums = [11, 13, 31]
Solution().minBitwiseArray(nums=nums)
```
