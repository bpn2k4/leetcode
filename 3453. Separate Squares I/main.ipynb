```python
from typing import List


class Solution:
  def separateSquares(self, squares: List[List[int]]) -> float:
    total_area = 0
    for square in squares:
      total_area += square[2] * square[2]
    haft = total_area / 2
    epsilon = 0.00001
    low, high = 0, 1000000001
    while (high - low) > epsilon:
      mid = (high + low) / 2
      area = 0
      for square in squares:
        if mid > square[1]:
          if mid <= (square[1] + square[2]):
            area += (mid - square[1]) * square[2]
          else:
            area += square[2] * square[2]
      if area >= haft:
        high = mid
      else:
        low = mid
    return low
```

```python
squares = [[0, 0, 1], [2, 2, 1]]
Solution().separateSquares(squares=squares)
```

```python
squares = [[0, 0, 2], [1, 1, 1]]
Solution().separateSquares(squares=squares)
```
