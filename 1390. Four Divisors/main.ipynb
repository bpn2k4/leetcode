```python
from typing import List


class Solution:
  def sumDivisors(self, num) -> int:
    n = -1
    find = False
    for i in range(2, int(num ** 0.5) + 1):
      if num % i == 0:
        if find:
          return 0
        if i * i == num:
          return 0
        find = True
        n = i
    if find:
      return 1 + num + n + num // n
    return 0

  def sumFourDivisors(self, nums: List[int]) -> int:
    total = 0
    mark = {}
    for num in nums:
      if num not in mark:
        mark[num] = self.sumDivisors(num)
      total += mark[num]
    return total
```

```python
nums = [21, 4, 7]
Solution().sumFourDivisors(nums=nums)
```

```python
nums = [21, 21]
Solution().sumFourDivisors(nums=nums)
```
