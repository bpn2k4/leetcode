```python
from typing import List


class Solution:
  def canPartitionGrid(self, grid: List[List[int]]) -> bool:
    m, n = len(grid), len(grid[0])
    total = sum(grid[i][j] for i in range(m) for j in range(n))
    if total % 2 != 0:
      return False
    t = 0
    for i in range(m):
      t += sum(grid[i])
      if t == total // 2:
        return True
      if t > total // 2:
        break
    t = 0
    for j in range(n):
      for i in range(m):
        t += grid[i][j]
      if t == total // 2:
        return True
      if t > total // 2:
        break
    return False
```

```python
grid = [[1, 4], [2, 3]]
Solution().canPartitionGrid(grid=grid)
```
