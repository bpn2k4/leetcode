```python
from typing import List


class Solution:
  def minPairSum(self, nums: List[int]) -> int:
    nums.sort()
    max_pair_sum = nums[0] + nums[-1]
    for i in range(1, len(nums) // 2):
      max_pair_sum = max(max_pair_sum, nums[i] + nums[-i - 1])
    return max_pair_sum
```

```python
nums = [3, 5, 2, 3]
Solution().minPairSum(nums=nums)
```

```python
nums = [3, 5, 4, 2, 4, 6]
Solution().minPairSum(nums=nums)
```
