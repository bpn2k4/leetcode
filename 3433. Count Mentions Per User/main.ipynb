```python
from typing import List


class Solution:
  def countMentions(self, numberOfUsers: int, events: List[List[str]]) -> List[int]:

    event2weight = {'OFFLINE': 0, 'MESSAGE': 1}
    events.sort(key=lambda x: (int(x[1]), event2weight[x[0]]))
    n = numberOfUsers
    output = [0] * n
    user_id2offline = [0] * n
    all_count = 0
    for event in events:
      if event[0] == 'OFFLINE':
        user_id2offline[int(event[2])] = int(event[1]) + 60
      else:
        if event[2] == 'ALL':
          all_count += 1
        else:
          if event[2] == 'HERE':
            for i in range(n):
              if user_id2offline[i] <= int(event[1]):
                output[i] += 1
          else:
            users = event[2].split(' ')
            for user in users:
              user_id = int(user[2:])
              output[user_id] += 1

    for i in range(n):
      output[i] += all_count

    return output
```

```python
numberOfUsers = 5
events = [["OFFLINE","28","1"],["OFFLINE","14","2"],["MESSAGE","2","ALL"],["MESSAGE","28","HERE"],["OFFLINE","66","0"],["MESSAGE","34","id2"],["MESSAGE","83","HERE"],["MESSAGE","40","id3 id3 id2 id4 id4"]]
Solution().countMentions(numberOfUsers=numberOfUsers, events=events)
```
