```python
from typing import List


class Solution:
  def largestMagicSquare(self, grid: List[List[int]]) -> int:
    m, n = len(grid), len(grid[0])

    row_ps = [[0] * (n + 1) for _ in range(m)]
    col_ps = [[0] * n for _ in range(m + 1)]

    diag_ps = [[0] * (n + 1) for _ in range(m + 1)]
    anti_ps = [[0] * (n + 2) for _ in range(m + 1)]

    for i in range(m):
      for j in range(n):
        row_ps[i][j + 1] = row_ps[i][j] + grid[i][j]
        col_ps[i + 1][j] = col_ps[i][j] + grid[i][j]
        diag_ps[i + 1][j + 1] = diag_ps[i][j] + grid[i][j]
        anti_ps[i + 1][j] = anti_ps[i][j + 1] + grid[i][j]

    for k in range(min(m, n), 1, -1):
      for i in range(m - k + 1):
        for j in range(n - k + 1):
          target = row_ps[i][j + k] - row_ps[i][j]
          valid = True

          for r in range(i, i + k):
            if row_ps[r][j + k] - row_ps[r][j] != target:
              valid = False
              break
          if not valid:
            continue

          for c in range(j, j + k):
            if col_ps[i + k][c] - col_ps[i][c] != target:
              valid = False
              break
          if not valid:
            continue

          if diag_ps[i + k][j + k] - diag_ps[i][j] != target or anti_ps[i + k][j] - anti_ps[i][j + k] != target:
            continue

          return k

    return 1
```

```python
grid = [[7, 1, 4, 5, 6], [2, 5, 1, 6, 4], [1, 5, 4, 3, 2], [1, 2, 7, 3, 4]]
Solution().largestMagicSquare(grid=grid)
```
