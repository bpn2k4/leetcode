```python
from typing import List


class Solution:
  def areSimilar(self, mat: List[List[int]], k: int) -> bool:
    m, n = len(mat), len(mat[0])
    k = k % n
    if k == 0:
      return True
    for i in range(m):
      if i % 2 == 0:
        for j in range(n):
          if mat[i][(j - k) % n] != mat[i][j]:
            return False
      else:
        for j in range(n):
          if mat[i][(j + k) % n] != mat[i][j]:
            return False
    return True
```

```python
mat = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
k = 4
Solution().areSimilar(mat=mat, k=k)
```

```python
mat = [[1, 2, 1, 2], [5, 5, 5, 5], [6, 3, 6, 3]]
k = 2
Solution().areSimilar(mat=mat, k=k)
```
