```python
from typing import List


class Solution:
  def countTrapezoids(self, points: List[List[int]]) -> int:
    
    MOD = 10**9 + 7
    y2count = {}
    for point in points:
      y = point[1]
      if y in y2count:
        y2count[y] += 1
      else:
        y2count[y] = 1
    
    total = 0
    prefix = 0
    for count in y2count.values():
      x = count * (count - 1) // 2
      total = (total + x * prefix) % MOD
      prefix = (prefix + x) % MOD
    
    return total
```

```python
points = [[1, 0], [2, 0], [3, 0], [2, 2], [3, 2]]
Solution().countTrapezoids(points=points)
```

```python
points = [[0, 0], [1, 0], [0, 1], [2, 1]]
Solution().countTrapezoids(points=points)
```
