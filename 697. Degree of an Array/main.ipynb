```python
from typing import List
from collections import Counter


class Solution:
  def findShortestSubArray(self, nums: List[int]) -> int:
    num2count = Counter(nums)
    max_frequency = max(num2count.values())
    smallest_length = len(nums)
    for num, count in num2count.items():
      if count == max_frequency:
        left = nums.index(num)
        right = len(nums) - 1 - nums[::-1].index(num)
        smallest_length = min(smallest_length, right - left + 1)
    return smallest_length
```

```python
nums = [1, 2, 2, 3, 1]
Solution().findShortestSubArray(nums=nums)
```

```python
nums = [1, 2, 2, 3, 1, 4, 2]
Solution().findShortestSubArray(nums=nums)
```
