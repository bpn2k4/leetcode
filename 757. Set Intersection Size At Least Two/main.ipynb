```python
from typing import List


class Solution:
  def intersectionSizeTwo(self, intervals: List[List[int]]) -> int:

    intervals.sort(key=lambda x: (x[1], -x[0]))
    count = 2
    n = len(intervals)
    left, right = intervals[0][1] - 1, intervals[0][1]
    for i in range(1, n):
      l, r = intervals[i]
      if left >= l:
        continue

      if l > right:
        count += 2
        left = r - 1
        right = r
      else:
        count += 1
        left = right
        right = r

    return count
```

```python
intervals = [[1, 3], [3, 7], [8, 9]]
Solution().intersectionSizeTwo(intervals=intervals)
```

```python
intervals = [[1, 3], [1, 4], [2, 5], [3, 5]]
Solution().intersectionSizeTwo(intervals=intervals)
```

```python
intervals = [[1, 2], [2, 3], [2, 4], [4, 5]]
Solution().intersectionSizeTwo(intervals=intervals)
```
