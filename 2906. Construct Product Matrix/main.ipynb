```python
from typing import List


class Solution:
  def constructProductMatrix(self, grid: List[List[int]]) -> List[List[int]]:
    m, n = len(grid), len(grid[0])
    p = m * n
    left_prefix = [1] * p
    right_prefix = [1] * p
    for i in range(p - 1):
      r, c = i // n, i % n
      left_prefix[i + 1] = left_prefix[i] * grid[r][c] % 12345
      j = p - i - 1
      r, c = j // n, j % n
      right_prefix[j - 1] = right_prefix[j] * grid[r][c] % 12345
    for i in range(m):
      for j in range(n):
        index = i * n + j
        grid[i][j] = (left_prefix[index] * right_prefix[index]) % 12345
    return grid
```

```python
grid = [[1, 2], [3, 4]]
Solution().constructProductMatrix(grid=grid)
```
