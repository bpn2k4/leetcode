```python
from typing import List


class Solution:
  def isTrionic(self, nums: List[int]) -> bool:
    if len(nums) < 4:
      return False
    n = len(nums)
    signal = []
    for i in range(1, n):
      if nums[i] == nums[i - 1]:
        return False
      if nums[i] > nums[i - 1]:
        signal.append(1)
      else:
        signal.append(-1)
    if signal[0] != 1 or signal[-1] != 1:
      return False
    change = 0
    for i in range(1, len(signal)):
      if signal[i] != signal[i - 1]:
        change += 1
    return change == 2
```

```python
nums = [1, 3, 5, 4, 2, 6]
Solution().isTrionic(nums=nums)
```
