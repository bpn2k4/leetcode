```python
from typing import List


class Solution:
  def getDescentPeriods(self, prices: List[int]) -> int:
    
    n = len(prices)
    prefix = [1] * n
    for i in range(1, n):
      if prices[i - 1] - prices[i] == 1:
        prefix[i] = prefix[i - 1] + 1

    return sum(prefix)
```

```python
prices = [3, 2, 1, 4]
Solution().getDescentPeriods(prices=prices)
```

```python
prices = [8, 6, 7, 7]
Solution().getDescentPeriods(prices=prices)
```
