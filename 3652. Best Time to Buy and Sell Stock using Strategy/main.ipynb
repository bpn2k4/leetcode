```python
from typing import List


class Solution:
  def maxProfit(self, prices: List[int], strategy: List[int], k: int) -> int:
    n = len(prices)
    k2 = k // 2
    sliding = sum(prices[i] * strategy[i] for i in range(n))
    max_profit = sliding
    for i in range(k2):
      sliding -= prices[i] * strategy[i]
    for i in range(k2, k):
      sliding -= prices[i] * strategy[i]
      sliding += prices[i]
    max_profit = max(max_profit, sliding)
    for i in range(k, n):
      sliding -= strategy[i] * prices[i]
      sliding += prices[i]
      sliding -= prices[i - k2]
      sliding += strategy[i - k] * prices[i - k]
      max_profit = max(max_profit, sliding)
    return max_profit
```

```python
prices = [4, 2, 8]
strategy = [-1, 0, 1]
k = 2
Solution().maxProfit(prices=prices, strategy=strategy, k=k)
```

```python
prices = [5, 4, 3]
strategy = [1, 1, 0]
k = 2
Solution().maxProfit(prices=prices, strategy=strategy, k=k)
```
