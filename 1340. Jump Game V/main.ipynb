```python
from typing import List


class Solution:
  def maxJumps(self, arr: List[int], d: int) -> int:
    visited = dict()

    def dfs(position: int):
      if position in visited:
        return
      visited[position] = 1
      i = position - 1
      while i >= 0 and position - i <= d and arr[position] > arr[i]:
        dfs(i)
        visited[position] = max(visited[position], visited[i] + 1)
        i -= 1
      i = position + 1
      while i < len(arr) and i - position <= d and arr[position] > arr[i]:
        dfs(i)
        visited[position] = max(visited[position], visited[i] + 1)
        i += 1

    for i in range(len(arr)):
      dfs(i)
    return max(visited.values())
```

```python
arr = [6, 4, 14, 6, 8, 13, 9, 7, 10, 6, 12]
d = 2
Solution().maxJumps(arr=arr, d=d)
```

```python
arr = [3, 3, 3, 3, 3]
d = 3
Solution().maxJumps(arr=arr, d=d)
```
