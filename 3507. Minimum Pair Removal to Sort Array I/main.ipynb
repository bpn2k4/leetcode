```python
from typing import List


class Solution:
  def minimumPairRemoval(self, nums: List[int]) -> int:
    
    def is_sorted(nums: List[int]) -> bool:
      if len(nums) <= 1:
        return True
      for i in range(len(nums) - 1):
        if nums[i] > nums[i + 1]:
          return False
      return True
    
    count = 0
    arr = nums.copy()
    while not is_sorted(arr):
      count += 1
      pair_sums = []
      for i in range(len(arr) - 1):
        pair_sums.append(arr[i] + arr[i + 1])
      minimum_pair_sum = min(pair_sums)
      for i in range(len(arr) - 1):
        if arr[i] + arr[i + 1] == minimum_pair_sum:
          arr[i] = minimum_pair_sum
          arr[i + 1] = -999999999
          break
      new_arr = []
      for num in arr:
        if num != -999999999:
          new_arr.append(num)
      arr = new_arr
    return count
```

```python
nums = [5, 2, 3, 1]
Solution().minimumPairRemoval(nums=nums)
```
