```python
from typing import List


class Solution:
  def maxSumDivThree(self, nums: List[int]) -> int:
    
    dp = [0, -1000000000, -1000000000]
    for num in nums:
      new_dp = dp.copy()
      for r in range(3):
        new_r = (r + num) % 3
        new_dp[new_r] = max(new_dp[new_r], dp[r] + num)
      dp = new_dp
    
    return dp[0]
```

```python
nums = [3, 6, 5, 1, 8]
Solution().maxSumDivThree(nums=nums)
```

```python
nums = [4]
Solution().maxSumDivThree(nums=nums)
```
