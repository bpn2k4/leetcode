```python
from typing import List


class Solution:
  MOD = 10**9 + 7
  pow10 = [1] * 100001
  for i in range(1, 100001):
    pow10[i] = pow10[i - 1] * 10 % MOD

  def sumAndMultiply(self, s: str, queries: List[List[int]]) -> List[int]:
    n = len(s)
    prefix_sum = [0] * (n + 1)
    x = [0] * (n + 1)
    cnt = [0] * (n + 1)
    for i, c in enumerate(s):
        d = int(c)
        prefix_sum[i + 1] = prefix_sum[i] + d
        x[i + 1] = (x[i] * 10 + d) % self.MOD if d > 0 else x[i]
        cnt[i + 1] = cnt[i] + (d > 0)
    m = len(queries)
    res = [0] * m
    for i in range(m):
        l = queries[i][0]
        r = queries[i][1] + 1
        length = cnt[r] - cnt[l]
        res[i] = (x[r] - x[l] * self.pow10[length]) * (prefix_sum[r] - prefix_sum[l]) % self.MOD
    return res
```

```python
s = "10203004"
queries = [[0, 7], [1, 3], [4, 6]]
Solution().sumAndMultiply(s=s, queries=queries)
```

```python
s = "1000"
queries = [[0, 3], [1, 1]]
Solution().sumAndMultiply(s=s, queries=queries)
```
