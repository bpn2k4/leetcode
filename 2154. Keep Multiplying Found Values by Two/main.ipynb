```python
from typing import List
from collections import Counter


class Solution:
  def findFinalValue(self, nums: List[int], original: int) -> int:
    
    num2count = Counter(nums)
    while original in num2count:
      original *= 2
    
    return original
```

```python
nums = [5, 3, 6, 1, 12]
original = 3
Solution().findFinalValue(nums=nums, original=original)
```

```python
nums = [2, 7, 9]
original = 4
Solution().findFinalValue(nums=nums, original=original)
```
