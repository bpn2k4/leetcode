```python
from typing import List


class Solution:
  def rangeAddQueries(self, n: int, queries: List[List[int]]) -> List[List[int]]:

    mat = [[0] * n for _ in range(n)]
    diff = [[0] * (n + 1) for _ in range(n + 1)]
    for x1, y1, x2, y2 in queries:
      diff[x1][y1] += 1
      diff[x2 + 1][y1] -= 1
      diff[x1][y2 + 1] -= 1
      diff[x2 + 1][y2 + 1] += 1

    for i in range(n):
      for j in range(n):
        if i > 0:
          diff[i][j] += diff[i - 1][j]

    for j in range(n):
      for i in range(n):
        if j > 0:
          diff[i][j] += diff[i][j - 1]

    for i in range(n):
      for j in range(n):
        mat[i][j] = diff[i][j]

    return mat
```

```python
n = 3
queries = [[1, 1, 2, 2], [0, 0, 1, 1]]
Solution().rangeAddQueries(n=n, queries=queries)
```
