```python
from typing import List


class Solution:
  def minimumDifference(self, nums: List[int], k: int) -> int:
    nums.sort()
    minimum_diff = nums[-1] - nums[0]
    for i in range(len(nums) - k + 1):
      minimum_diff = min(minimum_diff, nums[i + k - 1] - nums[i])
    return minimum_diff
```

```python
nums = [90]
k = 1
Solution().minimumDifference(nums=nums, k=k)
```

```python
nums = [9, 4, 1, 7]
k = 2
Solution().minimumDifference(nums=nums, k=k)
```
