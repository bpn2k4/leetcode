```python
from typing import List
from collections import Counter


class Solution:
  def isGood(self, nums: List[int]) -> bool:
    num2count = Counter(nums)
    n = len(nums) - 1
    for i in range(1, n + 1):
      if i not in num2count:
        return False
    if num2count[n] != 2:
      return False
    return True
```

```python
nums = [2, 1, 3]
Solution().isGood(nums=nums)
```

```python
nums = [1, 3, 3, 2]
Solution().isGood(nums=nums)
```

```python
nums = [1, 1]
Solution().isGood(nums=nums)
```

```python
nums = [3, 4, 4, 1, 2, 1]
Solution().isGood(nums=nums)
```
