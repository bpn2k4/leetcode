```python
from typing import List


class Solution:
  def canReach(self, arr: List[int], start: int) -> bool:
    n = len(arr)
    visited = [False] * n

    def dfs(i: int) -> bool:
      if arr[i] == 0:
        return True
      visited[i] = True
      left = i - arr[i]
      if left >= 0 and not visited[left]:
        if dfs(left):
          return True
      right = i + arr[i]
      if right < n and not visited[right]:
        if dfs(right):
          return True
      return False
    
    return dfs(start)
```

```python
arr = [4, 2, 3, 0, 3, 1, 2]
start = 5
Solution().canReach(arr=arr, start=start)
```

```python
arr = [4, 2, 3, 0, 3, 1, 2]
start = 0
Solution().canReach(arr=arr, start=start)
```
