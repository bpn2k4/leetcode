```python
from typing import List
import bisect


class Solution:
  def countNegatives(self, grid: List[List[int]]) -> int:
    m, n = len(grid), len(grid[0])
    count = 0
    for i in range(m):
      index = bisect.bisect_right(grid[i], 0, key=lambda x: -x)
      count += (n - index)
    return count
```

```python
grid = [[4, 3, 2, -1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]]
Solution().countNegatives(grid=grid)
```

```python
grid = [[3, 2], [1, 0]]
Solution().countNegatives(grid=grid)
```
