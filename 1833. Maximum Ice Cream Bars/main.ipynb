```python
from typing import List


class Solution:
  def maxIceCream(self, costs: List[int], coins: int) -> int:
    costs.sort()
    count = 0
    for cost in costs:
      if coins >= cost:
        coins -= cost
        count += 1
      else:
        break
    return count
```

```python
costs = [1, 3, 2, 4, 1]
coins = 7
Solution().maxIceCream(costs=costs, coins=coins)
```

```python
costs = [10, 6, 8, 7, 7, 8]
coins = 5
Solution().maxIceCream(costs=costs, coins=coins)
```
