```python
from typing import List


class Solution:
  def mapWordWeights(self, words: List[str], weights: List[int]) -> str:
    char2weight = dict(zip('abcdefghijklmnopqrstuvwxyz', weights))
    num2char = dict(zip(list(range(26))[::-1], 'abcdefghijklmnopqrstuvwxyz'))
    output = ''
    for word in words:
      total = sum([char2weight[char] for char in word])
      output += num2char[total % 26]
    return output
```

```python
words = ["abcd", "def", "xyz"]
weights = [5, 3, 12, 14, 1, 2, 3, 2, 10, 6, 6, 9, 7, 8, 7, 10, 8, 9, 6, 9, 9, 8, 3, 7, 7, 2]
Solution().mapWordWeights(words=words, weights=weights)
```
